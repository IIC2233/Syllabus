## üìö Sintaxis y patrones

| **Patr√≥n** | **Significado** | **Ejemplo** | **Resultado** |
|:-----------|:----------------|:-------------|:---------------|
| `.` | Cualquier car√°cter excepto salto de l√≠nea | `re.findall("c.t", "cat cot cut")` | `['cat', 'cot', 'cut']` |
| `^` | Inicio de la cadena | `re.findall("^Hola", "Hola mundo")` | `['Hola']` |
| `$` | Fin de la cadena | `re.findall("mundo$", "Hola mundo")` | `['mundo']` |
| `*` | 0 o m√°s repeticiones | `re.findall("ab*", "a ab abb abbb")` | `['a', 'ab', 'abb', 'abbb']` |
| `+` | 1 o m√°s repeticiones | `re.findall("ab+", "a ab abb abbb")` | `['ab', 'abb', 'abbb']` |
| `?` | 0 o 1 repetici√≥n | `re.findall("ab?", "a ab abb")` | `['a', 'ab', 'ab']` |
| `{n}` | Exactamente *n* repeticiones | `re.findall("a{2}", "caaat")` | `['aa']` |
| `{n,}` | Al menos *n* repeticiones | `re.findall("a{2,}", "caaat")` | `['aaa']` |
| `{n,m}` | Entre *n* y *m* repeticiones | `re.findall("a{1,3}", "aaaa")` | `['aaa']` |
| `[]` | Conjunto de caracteres posibles | `re.findall("[aeiou]", "regex")` | `['e', 'e']` |
| `[^ ]` | Cualquier car√°cter **no** incluido | `re.findall("[^0-9]", "a1b2c3")` | `['a', 'b', 'c']` |
| `\d` | D√≠gito (`[0-9]`) | `re.findall("\d+", "Edad: 18 a√±os")` | `['18']` |
| `\D` | No d√≠gito | `re.findall("\D+", "Edad: 18 a√±os")` | `['Edad: ', ' a√±os']` |
| `\w` | Car√°cter alfanum√©rico (`[A-Za-z0-9_]`) | `re.findall("\w+", "Hola mundo_123")` | `['Hola', 'mundo_123']` |
| `\W` | No alfanum√©rico | `re.findall("\W+", "Hola mundo_123!")` | `[' ', '!']` |
| `\s` | Espacio en blanco (espacio, tab, salto de l√≠nea) | `re.findall("\s", "Hola mundo")` | `[' ']` |
| `\S` | No espacio en blanco | `re.findall("\S+", "Hola mundo")` | `['Hola', 'mundo']` |
| `\|` | Alternativa l√≥gica (OR) | `re.findall("gato\|perro", "gato y perro")` | `['gato', 'perro']` |
| `()` | Grupo de captura | `re.findall("(ab)+", "abababc")` | `['ab', 'ab']` |


## ‚öôÔ∏è Funciones m√°s utilizadas del m√≥dulo `re`

| **Funci√≥n** | **Uso** | **Ejemplo** | **Resultado** |
|:-------------|:--------|:-------------|:---------------|
| `re.match(patron, texto)` | Busca el patr√≥n **solo al inicio** del texto | `re.match("Hola", "Hola mundo")` | Objeto `Match` |
| `re.search(patron, texto)` | Busca la **primera coincidencia** en cualquier parte del texto | `re.search("mundo", "Hola mundo")` | Objeto `Match` |
| `re.findall(patron, texto)` | Devuelve **todas las coincidencias** en una lista | `re.findall("\d+", "Hay 3 gatos y 4 perros")` | `['3', '4']` |
| `re.sub(patron, reemplazo, texto)` | Sustituye coincidencias por un texto nuevo | `re.sub("\d+", "X", "Hay 3 gatos y 4 perros")` | `"Hay X gatos y X perros"` |
| `re.split(patron, texto)` | Divide el texto usando el patr√≥n como separador | `re.split("\s+", "uno dos tres")` | `['uno', 'dos', 'tres']` |

Deber√°s buscar el camino m√°s corto desde S hasta T, considerando que no puedes pasar por X y para pasar por D necesitas una K (universal)

```python
import time
from collections import deque

# . camino
# X obst√°culo
# D puerta
# K llave
# S inicio
# T final 

laberinto = [
    list("S...XD"),
    list(".XX.X."),
    list(".X.KX."),
    list(".X.XX."),
    list("....DT"),
]

FILAS = len(laberinto)
COLUMNAS = len(laberinto[0])

for fila in range(FILAS):
    for columna in range(COLUMNAS):
        if laberinto[fila][columna] == "S":
            inicio = (fila, columna, False)
        elif laberinto[fila][columna] == "T":
            objetivo = (fila, columna)

def bfs(inicio, objetivo):
    cola = deque([[inicio]])
    visitados = {inicio}

    while cola:
        camino_actual = cola.popleft()
        fila_actual, columna_actual, tiene_llave = camino_actual[-1]

        laberinto[fila_actual][columna_actual] = "o" if laberinto[fila_actual][columna_actual] not in ("S","T") else laberinto[fila_actual][columna_actual]
        print("\nExplorando con BFS...")
        for fila in laberinto:
            print(" ".join(fila))
        time.sleep(0.5)

        if (fila_actual, columna_actual) == objetivo:
            return camino_actual

        for df, dc in [(1,0),(-1,0),(0,1),(0,-1)]:
            nueva_fila = fila_actual + df
            nueva_columna = columna_actual + dc

            if not (0 <= nueva_fila < FILAS and 0 <= nueva_columna < COLUMNAS):
                continue

            celda = laberinto[nueva_fila][nueva_columna]
            nuevo_estado_tiene_llave = tiene_llave

            if celda == "X":
                continue
            if celda == "D" and not tiene_llave:
                continue
            if celda == "K":
                nuevo_estado_tiene_llave = True

            nuevo_estado = (nueva_fila, nueva_columna, nuevo_estado_tiene_llave)

            if nuevo_estado not in visitados:
                visitados.add(nuevo_estado)
                cola.append(camino_actual + [nuevo_estado])

    return None

camino = bfs(inicio, objetivo)

if not camino:
    print("\nNo existe un camino hacia el tesoro.")
else:
    print("\nCAMINO M√ÅS CORTO ENCONTRADO (CON LLAVE):")
    print([(f,c) for (f,c,_) in camino])

    for (f,c,_) in camino[1:-1]:
        laberinto[f][c] = "*"

    print("\n=== MAPA FINAL ===")
    for fila in laberinto:
        print(" ".join(fila))
```

# üî¶ Ejercicio 2: Exploraci√≥n de Cuevas

Imaginemos una cueva, representada como grafo, donde cada nodo representa un espacio abierto, cada arista un t√∫nel.
Para asegurar una exploraci√≥n segura, debemos revisar si existen laberintos (ciclos), la profundidad m√°xima e ignorar aquellos t√∫neles bloqueados.

```python
cueva = {
    "Entrada": ["A", "B"],
    "A": ["C", "D"],
    "B": ["D", "E"],
    "C": ["F"],
    "D": ["F", "G"],
    "E": ["G"],
    "F": ["T√∫nel Final"],
    "G": ["T√∫nel Final"],
    "T√∫nel Final": []
}

tuneles_bloqueados = {("B", "E"), ("D", "G")}

def buscar_camino_mas_profundo(nodo, camino_actual):
    mejor_camino_encontrado = camino_actual

    for vecino in cueva[nodo]:
        if (nodo, vecino) in tuneles_bloqueados:
            continue

        if vecino not in camino_actual:
            nuevo_camino = buscar_camino_mas_profundo(vecino, camino_actual + [vecino])
            if len(nuevo_camino) > len(mejor_camino_encontrado):
                mejor_camino_encontrado = nuevo_camino

    return mejor_camino_encontrado


def camino_mas_profundo(origen):
    return buscar_camino_mas_profundo(origen, [origen])


camino_profundo = camino_mas_profundo("Entrada")

def detectar_ciclo(nodo, visitados, en_rama_actual):
    visitados.add(nodo)
    en_rama_actual.add(nodo)

    for vecino in cueva[nodo]:
        if (nodo, vecino) in tuneles_bloqueados:
            continue

        if vecino not in visitados:
            if detectar_ciclo(vecino, visitados, en_rama_actual):
                return True
        elif vecino in en_rama_actual:
            return True

    en_rama_actual.remove(nodo)
    return False


def existe_ciclo():
    return detectar_ciclo("Entrada", set(), set())

print("=== INFORME DE EXPLORACI√ìN DE LA CUEVA ===\n")

print("Camino m√°s profundo encontrado:")
print(" -> ".join(camino_profundo))
print(f"\nProfundidad m√°xima (en nodos): {len(camino_profundo) - 1}")

print("\n¬øLa cueva contiene laberintos (ciclos)?",
      "S√≠" if existe_ciclo() else "No")

print(f"\nT√∫neles bloqueados ignorados: {len(tuneles_bloqueados)}")
```

```python
import re

texto = "¬°Aaahhh! La cama de mi casa est√° ocupada por mi gata"
patron = r'a(s|m)*a'
coincidencias = re.findall(patron, texto)

print(f'Coincidencias encontradas: {coincidencias}')
print(f'Respuesta: {len(coincidencias)}')
```
