Un **Bundle** es un ítem de utilidad que permite agrupar hasta 64 objetos distintos en una sola ranura de tu inventario. En programación, el Bundle es el **Iterable** perfecto: un contenedor que guarda una colección (en este caso, una `deque` de tuplas con el nombre del ítem y su cantidad). El **Iterador**, por otro lado, es la "mano" que va rebuscando y extrayendo los objetos uno por uno. Una vez que sacas el último ítem, el saco queda vacío y, al intentar sacar otro, el sistema nos lanza un `StopIteration`

Deberás implementar un sistema donde un Bundle sea **Iterable** y posea su propio **Iterador** personalizado:

- `IterableBundle`: Representa el saco de ítems.

  - `__init__`: Recibe los ítems. 1. **`NameError`**: Si el usuario intenta usar `next()` sin haber creado el iterador primero.
