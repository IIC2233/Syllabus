## ¡Gato Multijugador! 😸

Luego de una productiva tarde estudiando los contenidos de la semana de Programación Avanzada, planeabas jugar con tu amigo Pepito, pero Steam está caído. Te encuentras aburrido en tu pieza, mirando al techo, cuando de pronto recuerdas: puedes usar un _socket_ para comunicarme con Pepito, e integrar una interfaz gráfica para crear un juego! Así es como te dispones a la tarea de crear un juego clásico en multijugador: el **gato**.

# Parte 1

Tienes programada gran parte del juego. Tienes un servidor, y un cliente con una interfaz gráfica separada entre frontend y backend. Sin embargo, te encuentras con un error muy extraño: cuando se unen dos jugadores, el segundo jugador no pasa de la ventana de inicio a la de juego como debería. Deberás analizar el código y completar las partes correspondientes para arreglar este error.

Hint: recuerda que estás trabajando con _threads_. ¿Cómo sabe el servidor que el cliente ya conectó las señales correspondientes a la interfaz gráfica?

# Parte 2

Ahora, te falta programar los turnos. De momento, tienes que el apretar una casilla está conectado al método `Cliente.jugar`. Tienes que lograr una comunicación con el servidor de la siguiente forma:

- `Cliente.jugar` debería informarle al servidor de la casilla apretada, sólo si es el turno del cliente y la casilla no está ocupada. Si logra jugar, deja de ser su turno.

- `Server.procesar_mensaje` procesará este mensaje, y usará el método `Partida.procesar_jugada` para ejecutar la jugada. Debe informarle del resultado de esto a ambos jugadores, y en caso de tener un ganador o un empate, eliminar la partida usando `Server.eliminar_partida`

- De vuelta en el cliente, `Cliente.procesar_mensaje` deberá primero revisar si se mandó un nuevo tablero, y actualizarlo si es el caso. Para eso deberás completar la función `Cliente.actualizar_tablero`.

- También en `Cliente.procesar_mensaje`, si la acción del mensaje es `"turno"`, significa que se ha cambiado de turno. Deberás actualizar el _label_ correspondiente y cambiar el atributo que representa al turno.
