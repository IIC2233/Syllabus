* **`acciones_juegos`**:

* **`obtener_generos`**:

* **`juegos_por_genero`**:
    * Debes cargar los juegos y filtrar la lista para devolver solo los nombres de aquellos que tengan el `genero` especificado en la URL.

* **`obtener_juego`**:
    * Debes encontrar un juego por su `id` o `nombre`.
    * Si el método es `GET`, lo devuelves.

* **`comparar_endpoint`**:
    * Debes obtener los datos de dos juegos e invocar `comparar_juegos(juego_1, juego_2)`, que se te entrega en `utils.py` y devuelve el juego con mejor `rating` (en caso de empate, el de mayor `horas_juego`). Declara el ganador devolviendo su nombre.

* **`obtener_condicional`**:
    * Debes filtrar la lista de juegos usando parámetros opcionales en la URL. Tienes que leer `genero`, `rating`, `horas_juego` y `precio` de `request.args` y devolver solo los juegos que cumplan con todos los criterios especificados (para `rating` y `horas_juego` el criterio es `>=`, y para `precio` es `<=`, ¡ojo con la dirección!). Como `genero` puede ser una lista de genéros, piensa en cómo la puedes mandar como un argumento en la URL.

Elabora una o más requests para cada uno de los endpoints solicitados.

````python
import requests
import json

BASE = "http://127.0.0.1:5001"
AUTH_TOKEN = "IIC2233"
````

### 1️⃣ GET `/juegos`

````python
response = requests.get(f"{BASE}/juegos")
print(response.status_code)
print(response.json())
````

### 2️⃣ POST `/juegos` (requiere autorización)

Crea un diccionario con los datos de un juego a agregar.

💡 *Nota:* El diccionario debe tener los campos `id` (int), `nombre` (str), `generos` (list[str]) y `stats` (dict con `rating` (int), `horas_juego` (int), `precio` (int))

````python
nuevo_juego = {
    "id": 999,
    "nombre": "hollow_knight_silksong",
    "generos": ["plataformas", "indie", "acción"],
    "stats": {"rating": 99, "horas_juego": 0, "precio": 20}
}

response = requests.post(f"{BASE}/juegos", data=json.dumps(nuevo_juego), headers={'Authorization': AUTH_TOKEN})
print(response.status_code)
print(response.json())
````
