# Programaci√≥n Funcional üöÄ

## üìê Paradigma

Paradigma de programaci√≥n en que todo es visto como el output de una funci√≥n, y cuyo valor depende **exclusivamente** del input (par√°metros de entrada). Veremos las funciones como flujos de transformaci√≥n de datos.

En este paradigma, todas las funciones son **funciones puras**:
- Para un mismo input siempre entregan el mismo output.
- No poseen **efectos secundarios** (modificar elementos externos al *scope* de la funci√≥n).

Consecuencias de trabajar con funciones puras:
1. Ya no se modifican elementos externos a una funci√≥n.  
   ‚û°Ô∏è Ahora, podemos descartar revisar funciones puras cuando tenemos que un valor o una estructura cambia inesperadamente.

2. Ahora el output solo depende del input.  
   ‚û°Ô∏è Esto permite que si el programa recuerda el input de una funci√≥n pura, no necesita llamar nuevamente a la funci√≥n.

3. Dada la ausencia de efectos secundarios, no podemos razonar mediante iteraciones como `for` o `while` que cambien las mismas variables iteraci√≥n tras iteraci√≥n (lo que implica un cambio de estado).

¬øEs la funci√≥n `anunciar_comida` pura?

````python
import random

minuta = {'Lunes': 'Lentejas', 'Martes': 'Milanesa',
          'Miercoles': 'Macarrones', 'Jueves': 'Jalape√±os rellenos',
          'Viernes': 'Vienesas', 'S√°bado': 'Sushi', 'Domingo': 'Donas c:'}

def anunciar_comida():
  dia = random.choice(list(minuta.keys()))
  return f'{dia} con {dia[0]} de {minuta[dia]}'

print(anunciar_comida())
````

¬øEs la funci√≥n `anunciar_comida_del_dia` pura?

````python
def anunciar_comida_del_dia(dia):
  return f'{dia} con {dia[0]} de {minuta[dia]}'

print(anunciar_comida_del_dia('Domingo'))
````

## üõµ Funciones Generadoras

- Funciones que retornan un generador.
- Entregan un valor utilizando `yield` y luego la ejecuci√≥n contin√∫a desde ese punto al volverse a llamar la funci√≥n.
- Para obtener sus valores puede iterarse sobre el generador retornado o puede utilizarse `next`.
- **Ventaja**: los valores son generados a medida que se van solicitando, evit√°ndose el uso innecesario de memoria.
- Ojo üëÄ: en Python basta que una funci√≥n contenga la sentencia `yield` para que esta sea generadora.

````python
dialogo1 = ['Hola!', 'funciones', 'y', 'desearte', '...']
dialogo2 = ['Somos', 'generadoras', 'queremos', 'mucha...', '...']

def f_generadora1(dialogo):
  yield from dialogo

def f_generadora2(dialogo):
  for linea in dialogo:
    yield linea

def not_generadora():
  return '\n¬°Yo tambi√©n soy generadora!'

f_gen1 = f_generadora1(dialogo1)
f_gen2 = f_generadora2(dialogo2)
for i in range(len(dialogo1)):
  if i == 4:
    print(not_generadora())
  print(next(f_gen1), end=' ')
  print(next(f_gen2), end=' ')
````

Podemos enviar valores a una funci√≥n generadora utiliando `send`. Antes de enviar un valor, la funci√≥n debe haber llegado en su ejecuci√≥n a la sentencia `yield`.

````python
guion = ['Buenas tardes,', 'os', 'deseo', 'mucho', 'exito', 'en', 'vuestro', 'estudio', '*reverencia*']

def recitador():
  pausa_dramatica = '...'
  linea = pausa_dramatica

  while True:
    print(linea)
    siguiente_linea = yield linea

    if siguiente_linea is None:
      linea = pausa_dramatica
    else:
      linea = siguiente_linea

recitador = recitador()
next(recitador)  # La ejecuci√≥n del recitador llega al yield
actual = 0
while actual < len(guion):
  if random.random() >= 0.5:
    recitador.send(guion[actual])  # Utilizamos send para enviar la siguiente l√≠nea del gui√≥n
    actual += 1
  else:
    next(recitador)
````

## üì† Funciones que retornan generadores

### - `map(f, iterable)`
Retorna el generador que resulta de aplicar la funci√≥n `f` a cada elemento del `iterable`.

`(f(x) for x in iterable)`

- Tambi√©n es posible hacer `map` de varios iterables de la forma `map(f, iterable1, iterable2, ...)`. El n√∫mero de iterables entregado a `map` debe ser igual a la cantidad de par√°metros que recibe la funci√≥n `f`.

Ejemplo:


````python
ciudades = ['Arica', 'Antofagasta', 'Santiago', 'Talca', 'Concepci√≥n', 'Punta Arenas']  # iterable
regiones = ('Regi√≥n de Arica y Parinacota', 'Regi√≥n de Antofagasta',
            'Regi√≥n Metropolitana', 'Regi√≥n del Maule', 'Regi√≥n del Bio-B√≠o',
            'Regi√≥n de Magallanes y de la Ant√°rtica Chilena')  # iterable

def ubicacion(ciudad, region):
  return f'{ciudad}, {region}, Chile'

ubicaciones = map(ubicacion, ciudades, regiones)  # elemento a elemento se van pasando a la funci√≥n

for ubicacion in ubicaciones:
  print(ubicacion)
````

### - `filter(f, iterable)`
Retorna el generador con los elementos del `iterable` tal que la funci√≥n `f` retorna `True`. Notar que `f` debe ser una funci√≥n booleana.

`(x for x in iterable if f(x))`

Ejemplo:

````python
fila = [('Pedro', 27), ('Juan', 32), ('Diego', 17)]

def es_mayor_de_edad(tupla):
  if tupla[1] < 18:
    print(f'Lo siento {tupla[0]}, no te puedo dejar pasar.')
  return tupla[1] >= 18

admitidos = filter(es_mayor_de_edad, fila)

for persona in admitidos:
  print(persona)
````

### - `reduce(f, iterable, initial: Any | None)`
Retorna el resultado de aplicar acumulativamente la funci√≥n `f` sobre los elementos del `iterable`.

`f(... f(f(elem1, elem2), elem3) ..., elemN)`

- Es posible entregar un valor inicial (`initial`) que ser√° el primer elemento procesado en el `reduce`.

Ejemplo:

````python
from functools import reduce

carrito_compra = ['zapatos', 'l√°mpara escritorio', 'jockey', 'aud√≠fonos']

def juntar_elementos(elem1, elem2):
  return f'{elem1}, {elem2}'

lista_de_compras = reduce(juntar_elementos,
                          carrito_compra[1:],  # iterable
                          carrito_compra[0])  # inicializador

print('****** TU LISTA DE COMPRAS ******')
print(lista_de_compras)
print('*********************************')
````

## üé≠ Funciones Lambda o funciones an√≥nimas

Una forma alternativa de definir funciones que nos ser√°n de utilidad a la hora de trabajar con `map`, `filter` y `reduce`. Caracter√≠sticas:
- Son definidas de forma an√≥nima, es decir, no poseen un nombre espec√≠fico.
- Son fugaces, lo que quiere decir que son utilizadas donde son definidas (se considera mala pr√°ctica guardar una funci√≥n `lambda` en una variable).

`lambda <par√°metros> : <valor a retornar>`

Ejemplo:

````python
ubicaciones = map(lambda ciudad, region : f'{ciudad}, {region}, Chile',
                  ciudades, regiones)  # ahora podemos definir in-place la funci√≥n entregada a map

for ubicacion in ubicaciones:
  print(ubicacion)
````

## Separaci√≥n Front-Back y Cambio de paradigma

Anteriormente teniamos un codigo muy **acoplado** y posiblemente de **baja cohesi√≥n**, es decir teniamos codigo muy dependiende de otro y posiblemente haciendo diversas tareas (respectivamente). Para esto surge el paradigma front-back el cual **propone una separaci√≥n de responsabilidades (Separation of Concerns)** dividiendo la arquitectura en dos componentes aut√≥nomos:

1.  **Frontend (Cliente):** Se encarga exclusivamente de la capa de presentaci√≥n y la experiencia de usuario (UI/UX).
2.  **Backend (Servidor):** Centraliza la l√≥gica de la aplicaci√≥n, concentrandose en calculos y manejo de logica.

## Ventanas

Las ventanas son el principal componente de PyQT en conjunto con botones, layouts, labels, entre otros. Los cuales **derivan de la clase base `QWidget` (o sus especializaciones como `QMainWindow`)** empleando herencia. Estos componentes se organizan visualmente mediante "Layouts" (Gestores de dise√±o) para adaptarse al tama√±o de la ventana y responden a las interacciones del usuario utilizando el mecanismo de comunicaci√≥n de Se√±ales y Eventos!


````python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QLabel, QVBoxLayout, QPushButton

class MiVentanaSimple(QWidget):
    def __init__(self):
        super().__init__()

        # Configuraci√≥n de la ventana
        self.setWindowTitle("Ventana Simple")
        self.setGeometry(100, 100, 300, 150)

        # Layout y Widgets
        layout = QVBoxLayout()
        etiqueta = QLabel("Ultima Ayudantia!!!")
        boton = QPushButton("Cerrar")

        layout.addWidget(etiqueta)
        layout.addWidget(boton)
        self.setLayout(layout)

if __name__ == "__main__":
    app = QApplication(sys.argv)
    ventana = MiVentanaSimple()
    ventana.show()
    sys.exit(app.exec_())
````

## Eventos y se√±ales

Para lograr comunicar todas las partes que est√°n ocurriendo a nivel front y back (y posiblemente con uso de los QThreads) es necesario una comunicaci√≥n front-back efectiva.

Para esto en PyQt  la forma en que los objetos en la aplicaci√≥n "hablan" entre s√≠ sin necesidad de conocer los detalles internos del otro.

Por ejemplo, un bot√≥n (`QPushButton`) maneja internamente los eventos crudos de "mouse presionado" y "mouse soltado". Cuando detecta que ambos ocurrieron sobre √©l, emite la se√±al **`clicked()`** la cual puede ser conectada y asi generar comportamientos!.

Lo m√°s importante es que **las se√±ales pueden transportar datos**. No solo avisan que algo pas√≥, sino que pueden enviar la informaci√≥n asociada. Si el usuario escribe en una caja de texto y presiona Enter (evento de tecla), la caja de texto puede emitir una se√±al que lleve consigo el **string** exacto que el usuario escribi√≥, envi√°ndolo desde el Front hacia donde sea necesario.

Para esto se plantea el siguiente esquema:

1.  Se pulsa un bot√≥n en el **FRONT**, o sea ocurre un **evento** que detona una se√±al ( `QPushButton.clicked()` ).
2.  Se env√≠a la **se√±al** de que se presion√≥ al **BACK** gracias al **main.py** (que conect√≥ la se√±al del front con una funci√≥n del back).
3.  El **BACK procesa** la l√≥gica asociada a ese bot√≥n.
4.  Env√≠a una **respuesta** (mediante otra se√±al propia que transporta los datos resultantes) al **FRONT** gracias al **main.py**.
5.  El **FRONT despliega** la respuesta recibida gracias al **BACK**.

De esta manera se desacopla la logica y la UI, logrando lo buscado en torno a acoplaci√≥n y cohesi√≥n!

````python
# Ejemplo Completado Simple sin back ni main pero mostrando eventos!

import sys
from PyQt5.QtWidgets import QApplication, QWidget, QLabel, QVBoxLayout, QPushButton

class MiVentanaSimple(QWidget):
    def __init__(self):
        super().__init__()

        # Configuraci√≥n de la ventana
        self.setWindowTitle("Ventana Simple")
        self.setGeometry(100, 100, 300, 150)

        # Layout y Widgets Para a√±adir
        layout = QVBoxLayout()
        etiqueta = QLabel("Hola mundo desde PyQt5")
        boton = QPushButton("Cerrar")

        # Conectar el bot√≥n a la acci√≥n de cerrar
        boton.clicked.connect(self.close)

        layout.addWidget(etiqueta)
        layout.addWidget(boton)
        self.setLayout(layout)

if __name__ == "__main__":
    app = QApplication(sys.argv)
    ventana = MiVentanaSimple()
    ventana.show()
    sys.exit(app.exec_())
````

# Threading üßµ

## üñ•Ô∏è Procesos, hilos, concurrencia y paralelismo

### Procesos e Hilos
Un proceso es simplemente un programa en ejecuci√≥n que tiene su propio espacio de memoria y sus propios recursos. Cada proceso funciona como un ‚Äúmundo aislado‚Äù (lo que pasa en uno no afecta a los dem√°s).

Dentro de un proceso pueden existir hilos (threads). Un hilo es un flujo independiente de instrucciones que avanza por su cuenta, pero a diferencia de los procesos, todos los hilos del mismo proceso comparten la misma memoria y los mismos recursos. Esto permite que colaboren f√°cilmente y que sean mucho m√°s livianos de crear que un proceso completo.

### Concurrencia
- Se ejecutan varias tareas que **avanzan en el tiempo de forma entrelazada**.
- Estas tareas no se ejecutan exactamente al mismo tiempo.
- Parece que ‚Äúvarias cosas pasan a la vez‚Äù porque el programa va alternando entre tareas muy r√°pido.

### Paralelismo
- Varias tareas se ejecutan **realmente al mismo tiempo**, por ejemplo en **n√∫cleos distintos del procesador**.
- Un programa puede ser concurrente sin ser paralelo y viceversa.

> En este curso usaremos la librer√≠a est√°ndar `threading` de Python, que nos permite programar **concurrencia con hilos dentro de un mismo proceso**.

## ‚öôÔ∏è Crear y manejar hilos con `threading.Thread`

Para usar hilos en Python:

````python
from threading import Thread

def funcion_a_ejecutar(arg1, arg2):
    ...

hilo = Thread(target=funcion_a_ejecutar, args=(arg1, arg2), daemon=False)
hilo.start()   # Comienza la ejecuci√≥n de un hilo (instancia de Thread guardada en la variable hilo)
hilo.join()    # Bloquea la ejecuci√≥n del hilo principal hasta que el hilo termine
````

- `target`: funci√≥n que se ejecutar√° en el hilo.
- `args`: tupla con los argumentos para `target`.
- `daemon` ‚Üí Si el valor es **True** el hilo se **termina autom√°ticamente** cuando el programa principal termina, si es **False** se sigue ejecutando.
- `start()` ‚Üí Comienza a ejecutar el c√≥digo de `funcion_a_ejecutar` en otro hilo, de manera concurrente.
- `join()` ‚Üí bloquea el hilo principal hasta que el hilo al cual se le hizo `join()` termine.

### Ejemplo: Lanzar varios hilos

````python
import time
from threading import Thread

def tarea(nombre, t):
    print(f"[{nombre}] Empezando, esperar√© {t} s")
    time.sleep(t)
    print(f"[{nombre}] Termin√©")

# Lanzamos 3 tareas concurrentes
hilos = [
    Thread(target=tarea, args=(f"Hilo-{i}", i))
    for i in (1, 2, 3)
]

for h in hilos:
    # Iniciamos todos los hilos uno por uno
    h.start()

for h in hilos:
    # Hacemos que el hilo principal espere a que todos los dem√°s terminen
    h.join()

# Imprimimos una vez que todos los hilos terminan
print("Todos los hilos han finalizado su ejecuci√≥n üí´")
````

## üîê Datos compartidos, condiciones de carrera y locks

Cuando varios hilos **modifican** el mismo dato al mismo tiempo, pueden aparecer **condiciones de carrera**.

Ejemplo: incrementar un contador compartido desde diferentes threads.

```python
contador += 1
```

Esa instrucci√≥n no es at√≥mica (no ocurre en un solo paso); internamente se descompone en tres pasos: leer el valor actual del contador, sumar uno y escribir el resultado. Si ambos hilos leen el mismo valor antes de que alguno lo actualice, podr√≠an terminar escribiendo valores incorrectos. Por ejemplo, si `contador` vale 10 y se quiere incrementar cada vez desde un thread diferente:

- Hilo A lee 10.
- Hilo B lee 10.
- A escribe 11.
- B escribe 11.

Resultado final: 11, aunque deber√≠an haberse realizado dos incrementos.

Para evitar estos problemas, existe el objeto `Lock`, que se crea mediante `threading.Lock()`. Un `Lock` asegura que solo un hilo a la vez pueda ejecutar la secci√≥n de c√≥digo donde se manipulan datos compartidos. Si un hilo toma el lock, los dem√°s deben esperar hasta que se libere.

M√©todos a utilizar sobre `lock = threading.Lock()`:

- `lock.acquire()`: el hilo intenta tomar el lock.
    - Si el lock est√° libre, lo toma inmediatamente.
    - Si otro hilo ya lo tiene, el hilo actual queda esperando hasta que se libere.
- `lock.release()`: libera el lock para que otro hilo pueda continuar.
    - Si no se llama a `release()` al finalizar la secci√≥n cr√≠tica, los dem√°s threads que lleguen a esta no podr√°n avanzar, ya que quedar√°n esperando a que se libere el lock.

Usando un lock alrededor de operaciones sensibles como actualizar un contador, garantizamos que solo un hilo a la vez ejecuta esa parte del c√≥digo, evitando condiciones de carrera.

### Ejemplo de condici√≥n de carrera

````python
import threading

NUM_THREADS = 5
ITERACIONES = 100

lock = threading.Lock()
with open('archivo.txt', 'w', encoding='utf-8') as arch:
    pass

def incrementar_con_lock(n_hilo, lock: threading.Lock):
    for i in range(ITERACIONES):
        lock.acquire()

        with open('archivo.txt', 'r', encoding='utf-8') as arch:
            lineas_anteriores = arch.readlines()

        with open('archivo.txt', 'w', encoding='utf-8') as arch:
            arch.write(f'Iteraci√≥n {i} del thread {n_hilo}\n')
            for linea in lineas_anteriores:
                arch.write(linea)

        lock.release()

hilos = [
    threading.Thread(target=incrementar_con_lock, args=(n_hilo, lock))
    for n_hilo in range(NUM_THREADS)
]

for h in hilos:
    h.start()

for h in hilos:
    h.join()
````

En este ejemplo, cada hilo lee el archivo y luego lo reescribe agregando su propia l√≠nea. Como estas operaciones **no son at√≥micas**, si dos hilos lo hacen al mismo tiempo pueden sobrescribir el trabajo del otro y perder informaci√≥n. El `Lock` evita esto asegurando que solo un hilo acceda al archivo en cada momento.

## üéå Eventos (`threading.Event`)
Un `Event` es una herramienta de sincronizaci√≥n que permite coordinar hilos mediante una se√±al compartida. Funciona como un interruptor que puede estar en dos estados: apagado (`False`) o encendido (`True`).
- Un `Event` se puede crear como `event = threading.Event()` y este comienza **apagado**.
- Un hilo puede utilizar `event.wait()`:
    - Si el evento est√° **apagado** el hilo quedar√° bloqueado esperando a que se encienda.
    - Si el evento est√° **encendido** el hilo contin√∫a con el c√≥digo que sigue.
- Otro hilo puede **encender** el evento utilizando `event.set()`
- Se puede **apagar** el evento utilizando `event.clear()`

Los eventos son √∫tiles cuando queremos que un hilo espere una condici√≥n externa antes de continuar su ejecuci√≥n, o cuando necesitamos detener hilos de manera controlada sin forzar su cierre. En otras palabras, permiten coordinar el momento exacto en que ciertos hilos deben avanzar.

### Ejemplo coordinar inicio de trabajadores

````python
import threading
import time
import random

# Evento que controla el inicio
inicio_evento = threading.Event()

def trabajador(id_hilo: int, evento: threading.Event):
    print(f"Hilo {id_hilo} listo y esperando...")

    evento.wait()  # Cada hilo queda detenido hasta que la se√±al se active

    # Cuando la se√±al se activa, todos comienzan aqu√≠
    print(f"Hilo {id_hilo} comenz√≥ a trabajar.")
    time.sleep(random.uniform(0.5, 1.5))
    print(f"Hilo {id_hilo} termin√≥.")

# Crear hilos
hilos = [
    threading.Thread(target=trabajador, args=(i, inicio_evento))
    for i in range(5)
]

# Iniciar hilos (quedan esperando)
for h in hilos:
    h.start()

time.sleep(1)
print("\nTodos los hilos est√°n listos. Lanzando se√±al...\n")

# Activar evento para que todos empiecen a la vez
inicio_evento.set()

# Esperar a que terminen
for h in hilos:
    h.join()

print("\nTodos los hilos han terminado.")

````

En este ejemplo, varios hilos se preparan y quedan detenidos usando `event.wait()`. Todos comienzan a trabajar al mismo tiempo cuando el hilo principal activa la se√±al con `event.set()`, permitiendo sincronizar el inicio de m√∫ltiples tareas concurrentes.

## üåê APIs, Webservices y Endpoints

Una **API** (Application Programming Interface) es un conjunto de reglas que define c√≥mo un software puede comunicarse con otro.  
Cuando esta comunicaci√≥n ocurre mediante **HTTP**, hablamos de una **API web** o **webservice**.

Para interactuar con esos recursos se utilizan **m√©todos HTTP** como `GET`, `POST`, `PUT`, `PATCH` o `DELETE`. Estos le indican a la API que tipo de consulta debe hacer sobre una url definida. En el curso las APIs se crean utilizando el m√≥dulo `flask`

## üéØEndpoints
Est√°n definidos por:

### 1. M√©todo HTTP (la acci√≥n que quiero realizar)
- `GET`: obtener datos.
- `POST`: crear un recurso nuevo.
- `PUT`: reemplazar **completamente** un recurso existente (sobrescribe todos los datos).
- `PATCH`: modificar **parcialmente** un recurso existente (solo actualiza los campos enviados).
- `DELETE`: eliminar un recurso.

### 2. Path del recurso
- Asociamos una funci√≥n(c√≥digo a ejecutar) cuando se hace una request a cada url.
- Ejemplo: `/usuarios/<id>`
- Puede incluir **par√°metros de ruta** (por ejemplo, el ID del usuario).

### 3. Par√°metros de consulta (query params)
Son opcionales y van despu√©s de `?`: `/usuarios?activo=true&limite=10`. Se pueden obtener con `flask.request.args.get("activo")` y `flask.request.args.get("limite")`

### 4. Cuerpo de la request (body)
Generalmente se env√≠a en **JSON**, indicando los datos a crear o modificar. Se puede obtener con `flask.request.get_json()`

### 5. Respuesta del servidor (response)
La respuesta que entrega un servidor suele representarse mediante el objeto **`flask.Response()`**, el cual contiene:

- un **c√≥digo de estado** (status=) HTTP (por ejemplo 200, 201, 400, 404, 500‚Ä¶)
- un **cuerpo** (response=), generalmente JSON.
- el **tipo de respuesta** (content_type=), generalmente `application/json`

### Ejemplo de API
Api con dos endpoints con m√©todos `GET` y `POST` respectivamente.

````python
from flask import Flask, request, Response, json

app = Flask(__name__)

usuarios = []
id_usuario = [0]

# GET: obtener un usuario
@app.route("/usuarios/<int:user_id>", methods=["GET"])
def obtener_usuario(user_id):
    for usuario in usuarios:
        if usuario['id'] == user_id:
            return Response(
                response=json.dumps(usuario),
                status=200,
                content_type="application/json"
            )

# POST: crear un usuario
@app.route("/usuarios", methods=["POST"])
def crear_usuario():
    body = request.get_json() # Obtener JSON del body
    id_usuario[0] += 1

    nuevo_usuario = {
        "id": id_usuario[0],
        "nombre": body.get("nombre", "Sin nombre")
    }

    usuarios.append(nuevo_usuario)

    return Response(
        response=json.dumps({'mensaje': f'Usuario {nuevo_usuario} creado con id: {id_usuario}'}),
        status=201, # Recurso creado
        content_type="application/json"
    )

if __name__ == "__main__":
    app.run(host="localhost", port=5001)
````

## Sockets

Un _socket_ es un elemento que nos permite transmitir datos desde un dispositivo a otro. Un servidor y un cliente lo usan de la siguiente forma:

- Servidor
  - Crea el socket asociado al servidor.
  - Hace `bind` con su direcci√≥n IP y el puerto desde donde escuchar√°.
  - Hace `listen` para empezar a escuchar. Se le puede indicar una cantidad m√°xima de clientes conectados.
  - Hace `accept`, que queda esperando hasta que se conecte un nuevo cliente.
  - Hace `recv` y `send` con **el socket del cliente** para recibir y mandar informaci√≥n como bytes.

- Cliente
  - Crea el socket asociado al cliente.
  - Hace `connect` con la direcci√≥n IP y el puerto **del servidor**.
  - Hace `recv` y `send` con **su socket** para recibir y mandar informaci√≥n como bytes.

Para un servidor es fundamental el uso de _threads_ en conjunto con networking para mantener la comunicaci√≥n simult√°nea con todos los clientes. Asimismo, el cliente generalmente tiene un _thread_ que se encarga de escuchar al servidor en todo momento.

- Arquitectura P2P: es un tipo de arquitectura alternativa a Cliente-Servidor. Cada usuario es a la vez cliente y servidor y se comunican de a pares (peer to peer). Es una arquitectura descentralizada ideal para cosas como transferencia de archivos (üè¥‚Äç‚ò†Ô∏è).

## Recorrido de grafos

### BFS (Breadth First Search)

La b√∫squeda en amplitud recorre el grafo por niveles, utilizando una cola que agrega los nodos (FIFO).

Ideal para b√∫squeda de caminos m√°s cortos.

### DFS (Depth First Search)

La b√∫squeda en profundidad recorre el grafo por un camino hasta que no queden m√°s nodos, utilizando un stack que agrega los nodos. (LIFO)

````python
def dfs(grafo, inicio):
    # Vamos a mantener un set con los nodos visitados.
    visitados = set()

    # El stack de siempre, comienza desde el nodo inicio.
    stack = [inicio]

    while len(stack) > 0:
        vertice = stack.pop()
        # Detalle clave: si ya visitamos el nodo, ¬°no hacemos nada!
        if vertice in visitados:
            continue

        # Lo visitamos
        print(vertice)
        visitados.add(vertice)

        # Agregamos los vecinos al stack si es que no han sido visitados.
        for vecino in grafo[vertice]:
            if vecino not in visitados:
                stack.append(vecino)

    return list(visitados)


from collections import deque

def bfs(grafo, inicio):
    # Vamos a mantener un set con los nodos visitados.
    visitados = set()

    # Usamos una cola, comienza desde el nodo inicio
    queue = deque([inicio])

    while len(queue) > 0:
        vertice = queue.popleft()
        # Detalle clave: si ya visitamos el nodo, ¬°no hacemos nada!
        if vertice in visitados:
            continue

        # Lo visitamos
        print(vertice)
        visitados.add(vertice)

        # Agregamos los vecinos a la cola si es que no han sido visitados
        for vecino in grafo[vertice]:
            if vecino not in visitados:
                queue.append(vecino)

    return list(visitados)
````
