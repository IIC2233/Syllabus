## Introducción

### DCCgram, el drama de las señales

En el `DCC` se lanzó una nueva red social interna: **DCCgram**, donde ayudantes y profes suben perfiles e historias para compartir recomendaciones semanales. El problema es que el sistema quedó a medio cocinar: las ventanas existen, pero **nadie se está hablando correctamente entre sí**, y además **faltan vistas clave de perfiles e historias**.
    
Tu misión en esta ayudantía es arreglar el cahuín técnico conectando señales, separando frontend y backend, completando esas vistas, y dejando que la clave de acceso haga su pega sin romper la app. Si todo sale bien, la comunidad podrá navegar perfiles e historias sin que explote PyQt en plena ayudantía.


### Descripción del programa
#### Frontend
En el directorio `frontend/` de definen 4 clases de ventanas: `VentanaInicio`, `VentanaPerfiles`, `VentanaPerfil` y `VentanaHistoria`, y todas son gestionadas por una clase contenedora `Frontend`. `VentanaInicio` se encarga de recibir la clave de acceso y reenviarla, `VentanaPerfiles` muestra los perfiles disponibles, `VentanaPerfil` muestra la información de un perfil individual, y `VentanaHistoria` muestra las historias asociadas a ese perfil. Además, `VentanaPerfiles`, `VentanaPerfil` y `VentanaHistoria` permiten navegar de vuelta a la ventana anterior.

Importante: `VentanaInicio` y `VentanaPerfiles` ya se encuentran implementadas y **no se deben modificar**.

La clase `Frontend` instancia y coordina las ventanas, recibe sus señales, reenvía al backend la clave ingresada, y recibe desde el backend el resultado de la validación para decidir la navegación. En particular, maneja una colección de `VentanaPerfil` y una colección de `VentanaHistoria`, una por usuario y por historia respectivamente.

#### Backend
En `backend/backend.py` existe la clase `Logica`, que administra un archivo `clave_acceso.txt`, registra la clave cuando esta no existe, la valida cuando sí existe, e informa al frontend el resultado de la validación **mediante señales**.


#### Main
Finalmente en `main.py` se inicializan las instancias y se conectan las señales backend <-> frontend.

    - Además deberás completar los siguientes métodos:

          - Leer la clave guardada desde el archivo en `self.ruta_clave`, y compararla con la clave recibida. Si coinciden, emitir `senal_resultado_validacion` con `True` y el mensaje `Acceso correcto.`, si no coinciden, emitir `senal_resultado_validacion` con `False` y el mensaje `Clave incorrecta.`.
