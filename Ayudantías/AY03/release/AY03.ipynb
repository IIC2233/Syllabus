## Introducci√≥n

En esta ayudant√≠a partiremos viendo el concepto de **herencia** en OOP. La herencia es una forma de reciclar c√≥digo en la definici√≥n de clases. Si queremos crear una clase nueva que se parece a una que ya tenemos y se le agregan funcionalidades, podemos hacer que la nueva clase herede de la ya creada para que mantenga todos sus m√©todos y atributos. En los ejercicios veremos ejemplos de esto.
