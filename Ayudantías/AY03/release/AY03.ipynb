## Introducción

En esta ayudantía partiremos viendo el concepto de **herencia** en OOP. La herencia es una forma de reciclar código en la definición de clases. Si queremos crear una clase nueva que se parece a una que ya tenemos y se le agregan funcionalidades, podemos hacer que la nueva clase herede de la ya creada para que mantenga todos sus métodos y atributos. En los ejercicios veremos ejemplos de esto.

Las **properties** son funcionalidades en python que nos permiten, en una clase, tener datos que se acceden como atributos pero por detrás funcionan como métodos. Una **property** siempre debe tener un **getter**, que sirve para acceder a ella. Adicionalmente, pueden tener un **setter**, que sirve para modificarla.
