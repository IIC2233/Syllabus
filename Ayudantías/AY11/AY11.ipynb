### Sobre `TypeError` y `ValueError`

Todo `TypeError` es un `ValueError`, ya que los problemas de tipo de una variable provienen inherentemente del valor que esta posea. Pero no todo `ValueError` es un `KeyError`.

La diferencia existe ya que siempre es conveniente tener información más específica respecto a una excepción.


## Levantamiento de excepciones

Python nos permite levantar excepciones utilizando la sentencia `raise`. Podemos definir un comportamiento esperado y, si el código se sale de este comportamiento, levantar una excepción de manera personalizada.

````python
def convert_coordinate(coordinate_string):
    # Si el input no es del tipo esperado
    if not isinstance(coordinate_string, str):
        # aquí se genera una excepción y se incluye información para el usuario.
        raise TypeError("Coordinate must be a string")

    # Si el input no sigue el formato esperado
    if "," not in coordinate_string:
        # aquí se genera otra excepción y se incluye información para el usuario.
        raise ValueError("Coordinate must be separated by a comma")

    coord_x, coord_y = coordinate_string.split(",")
    return (int(coord_x), int(coord_y))
````

## Manejo de Excepciones: `try` y `except`

Cada vez que se levanta una excepción durante la ejecución del código es posible atraparla mediante el uso de `try` y `except`.

La sentencia `try`: define un bloque de código (*scope*). Si se levanta una excepción dentro del *scope*, la excepción es capturada. Luego debe existir una (o más) sentencias `except` que implementan el comportamiento para manejar la excepción capturada.

Al momento de que se captura la excepción dentro de un `try`, inmediatamente se salta al bloque de código de uno de los `except`. Una vez se ejecuta ese bloque, se continua la ejecución del programa y **NO** se regresa a la sentencia que gatillo la excepción.

````python
# 1)

# No se atrapa ninguna excepción, por lo tanto no se previene un posible error.
def divide(num, den):
    # Esta función terminará el programa cuando el
    # denominador den sea 0.
    return float(num) / float(den)

# cambia esto por 0 para ver el comportamiento
r = 4
try:
    # Dentro de este bloque ejecutamos el código que PODRÍA
    # arrojar una excepción.
    print(divide(5, r))
    print("I already did the division.")

except (ZeroDivisionError) as error:
    # Aquí manejamos la excepción que pueda ser lanzada en
    # el bloque anterior. Si un error del tipo ZeroDivisonError
    # ocurre, se ejecuta este bloque y el resto del programa
    # continúa su ejecución normal. La excepción, como objeto,
    # se puede acceder con la variable error.
    print(f"Error: {error}")
    print("How could you think of dividing by zero? Why are you like this?")

print("The program continues after the try/except")
````

````python
# 2) Manejo de múltiples excepciones

def convert_coordinate(coordinate_string):
    # Verificamos que el input es del tipo esperado.
    if not isinstance(coordinate_string, str):
        # Intencionalmente levantamos la excepción. Recuerda que las excepciones son objetos.
        raise TypeError("Coordenada debe ser un string")

    # Verificamos que el input sigue el formato esperado.
    if "," not in coordinate_string:
        # El mensaje incluido en la excepción es el que se despliega
        # cuando la manejamos después.
        raise ValueError("Coordenada debe estar separada por una coma")

    coord_x, coord_y = coordinate_string.split(",")
    return (int(coord_x), int(coord_y))


# Notamos que para un solo try, podemos definir más de un scope except
try:
    print(convert_coordinate([42, 3]))

# En esta parte manejamos las excepciones una vez que son lanzadas.
except TypeError as error:
    # Este bloque sólo maneja excepciones del tipo TypeError.
    print(f"Error: {error}")
    print("Check the argument type.")

except ValueError as error:
    # Este bloque sólo maneja excepciones del tipo ValueError.
    print(f"Error: {error}")
    print("A ValueError occurred. Please verify your values.")
````

Respecto a los `except`, podemos definir una por tipo de excepción, así como también capturar múltiples tipos de excepciones dentro de un bloque `except`.

## Flujos complementarios: `else` y `finally`

- `else`: Define un *scope* de intrucciones que se ejecutarán **siempre y cuando no se haya lanzado ninguna excepción**.
- `finally`: Define un *scope* van instrucciones que se realizan **siempre, independientemente si ocurrió una excepción o no**. Suele ser utilizado para ejecutar funciones de limpieza, como por ejemplo, el cierre de un archivo después de ser procesado.

### `finally` y `return`

Un bloque `finally` SIEMPRE se ejecuta, incluso si dentro del bloque `try` (o `except`) se encuentra un `return`

Dentro de una función, cuando se llega a un `return` dentro de un `try` (o `except`):
1. se almacena el valor del `return`
2. inmediatamente se salta al bloque `finally`
3. ejecuta el código y solamente cuando este bloque se acaba, retorna el valor almacenado del paso 1.


````python
def check_finally():
    try:
        return "Retorna desde el try"
    finally:
        print("Esto imprime ANTES del return")

check_finally()
````

# Iterables e Iteradores

## Iterable

Un **iterable** es un objeto que puede recorrerse con `for` (por ejemplo listas, sets, tuplas, diccionarios, strings).

Un objeto es iterable si cumple al menos una de estas condiciones:
- Implementa `__iter__()` y retorna un iterador.
- Implementa `__getitem__()` con índices válidos y, cuando ya no hay elementos, levanta `IndexError`.

El `for` necesita obtener elementos uno a uno. Si el objeto no entrega ese comportamiento, no se puede recorrer.

````python
# 1) Iterable simple (built-in)
numeros = [10, 20, 30]

for n in numeros:
    print(n)
````

````python
# 2) Iterable propio usando __getitem__
class DobleCaracter:
    def __init__(self, texto: str):
        self.texto = texto

    def __getitem__(self, i: int):
        # str ya levanta IndexError cuando i no existe
        return self.texto[i] * 2


it = DobleCaracter("py")
for x in it:
    print(x)  # pp, yy
````

## Iterador
Un **iterador** es el objeto que lleva el estado del recorrido, en otras palabras, un iterador en un objeto que itera sobre el iterable.

Debe implementar:
- `__next__()` para entregar el siguiente elemento.
- `__iter__()` retornando `self` (para que también sea usable en `for`).

Cuando se acaban los elementos, `__next__()` debe levantar `StopIteration`.

Un **iterable** se puede recorrer múltiples veces (cada `for` pide un iterador nuevo). En cambio, un **iterador** normalmente se consume una sola vez.

````python
# 3) Iterable + Iterador explícitos (__iter__ y __next__)
class RangoSimple:
    def __init__(self, inicio: int, fin: int):
        self.inicio = inicio
        self.fin = fin

    def __iter__(self):
        return IteradorRango(self.inicio, self.fin)


class IteradorRango:
    def __init__(self, actual: int, fin: int):
        self.actual = actual
        self.fin = fin

    def __iter__(self):
        return self

    def __next__(self):
        if self.actual >= self.fin:
            raise StopIteration
        valor = self.actual
        self.actual += 1
        return valor


for n in RangoSimple(3, 7):
    print(n)  # 3 4 5 6
````

````python
# 4) Consumir un iterador manualmente con next()
valores = [5, 6, 7]
it = iter(valores)

print(next(it))  # 5
print(next(it))  # 6
print(next(it))  # 7
# print(next(it))  # StopIteration
````

````python
# 5) Un iterador se agota; el iterable se puede recorrer de nuevo
datos = [1, 2, 3]
it = iter(datos)

print(list(it))  # [1, 2, 3]
print(list(it))  # [] (agotado)

print(list(datos))  # [1, 2, 3]
print(list(datos))  # [1, 2, 3]
````

## Generadores

Un **generador** es un caso especial de iterador. Se crea, por ejemplo, con comprensión usando paréntesis:
`gen = (expresion for x in datos)`

Y cuenta con las siguientes características:
- Se consumen una sola vez.
- Funcionan con `for` y con `next()`.
- Ahorran memoria: generan valores bajo demanda y no almacenan toda la secuencia como una lista.

Diferencia importante en comprensión de generadores:
- `(... for x in datos if condicion)` -> **filtra** elementos.
- `(... if condicion else otro for x in datos)` -> **transforma** cada elemento.

# Programación Funcional

## Paradigma Funcional

Python permite escribir código funcional, combinándolo con otros paradigmas. En este enfoque pensamos el programa como una **transformación de datos**: cada función recibe un estado y devuelve otro.

Conceptos centrales:
- **Funciones puras**: para el mismo input, entregan el mismo output.
- **Sin efectos secundarios**: la función no modifica estado externo (variables globales, archivos, consola, etc.).
- **Inmutabilidad lógica**: en vez de mutar estructuras, se construyen nuevas versiones.
- **Funciones de primera clase**: las funciones se pueden guardar en variables, pasar como argumento y retornar desde otras funciones.

Cuando una función es pura, es más fácil de razonar, reutilizar y optimizar (por ejemplo, con `@cache`).


## Funciones Generadoras

Una **función generadora** es una función que contiene `yield`.

Diferencias con una función tradicional:
- `return` termina la función inmediatamente.
- `yield` entrega un valor y **pausa** la ejecución, guardando su estado interno para continuar después.

Esto permite producir datos de forma incremental, evitando cargar todo en memoria a la vez.

Ideas clave:
- Al llamar la función, se obtiene un **generador** (no se ejecuta todo inmediatamente).
- Se puede recorrer con `for` o con `next()`.
- Un generador se consume una vez.
- `yield from` delega la iteración a otro iterable/generador.


````python
def pares_hasta(n: int):
    for i in range(n + 1):
        if i % 2 == 0:
            yield i

gen = pares_hasta(10)
print(next(gen))
print(next(gen))
print(list(gen))  # continúa desde donde quedó

def generador_hijo():
    yield from range(3)

def generador_padre():
    yield -1
    yield from generador_hijo()
    yield 99

print(list(generador_padre()))
````

## Funciones que trabajan con iterables

### Que necesitan de una función
- `map(f, iterable)`: aplica `f` a cada elemento.
- `filter(f, iterable)`: conserva elementos donde `f(elem)` es `True`.
- `reduce(f, iterable[, inicial])`: acumula el iterable a un solo resultado.

#### Funciones Lambda o funciones anónimas

Una forma alternativa de definir funciones que nos serán de utilidad a la hora de trabajar con `map`, `filter` y `reduce`. Características:
- Son definidas de forma anónima, es decir, no poseen un nombre específico.
- Son fugaces, lo que quiere decir que son utilizadas donde son definidas (se considera mala práctica guardar una función `lambda` en una variable).

`lambda <parámetros> : <valor a retornar>`

### Que no necesitan de una función

Funciones que encajan muy bien con el paradigma funcional porque operan directamente sobre iterables.

| Función       | Modo de uso                                         | Módulo        | Explicación breve                                                                 |

````python
from functools import reduce

numeros = [1, 2, 3, 4, 5, 6]

# Necesitan una función
cuadrados = list(map(lambda x: x ** 2, numeros))
pares = list(filter(lambda x: x % 2 == 0, numeros))
suma_total = reduce(lambda acc, x: acc + x, numeros, 0)

print("Cuadrados:", cuadrados)
print("Pares:", pares)
print("Suma total:", suma_total)


# No necesitan una función
print("\nEnumerate:")
for i, valor in enumerate(["a", "b", "c"], start=1):
    print(i, valor)

print("\nZip:")
nombres = ["Ana", "Luis", "Mora"]
edades = [21, 20, 23]
for par in zip(nombres, edades):
    print(par)

print("\nall/any:")
print(all(x > 0 for x in numeros))
print(any(x > 5 for x in numeros))
````

# Threading

Muchas veces al programar necesitaremos hacer dos cosas al mismo tiempo, y para eso podemos usar la librería `threading`. Esta nos permite programar *concurrencia*. Es importante conocer la diferencia entre:


- **Concurrencia**: gestionar varias tareas de forma intercalada en un mismo núcleo o procesador. Aunque solo una tarea se ejecute en un instante dado, los cambios de contexto se realizan tan rápido que aparentan simultaneidad. Se usa la librería `threading`.
- **Paralelismo**: ejecución simultánea de varias tareas en diferentes núcleos o procesadores. Se usan otras librerías.

## Thread
Un `Thread` es un objeto que permite ejecutar un trozo de código de manera concurrente al hilo principal. Podemos definirlos con una función objetivo o hacer una clase que herede de `Thread` y sobreescriba el método `run`.

````python
from threading import Thread
import random, time

def contar(n, paridad):

    for i in range(n):
        if i % 2 == paridad:
            time.sleep(random.randint(0, 1))
            print(i)


thread_1 = Thread(target=contar, args=(10, 0))
thread_2 = Thread(target=contar, args=(10, 1))

thread_1.start()
thread_2.start()

# al parecer algunos números se imprimen en la misma línea, cómo arreglamos eso? 🤔
````

## join
Si estamos en la `thread_1` y ejecutamos `thread_2.join()`, vamos a parar la ejecución de la `thread_1` hasta que termine `thread_2`.

````python
class MiThread(Thread):

    def __init__(self, numero):
        super().__init__()
        self.numero = numero

    def guardar(self, otra_thread):
        self.otra_thread = otra_thread
        
    def run(self):
        print(f"Thread {self.numero}: Voy a contar!")
        for i in range(10):
            print(f"Thread {self.numero}: {i}")
            time.sleep(0.5 * self.numero)

            # la thread 1 espera a la thread 2
            if i == 5 and self.numero == 1:
                print(f"Thread {self.numero}: esperaré a la otra thread para seguir")
                self.otra_thread.join()
        print(f"Thread {self.numero}: Terminé :v")


thread_1 = MiThread(1)
thread_2 = MiThread(2)
thread_1.guardar(thread_2)
thread_2.guardar(thread_1)

thread_1.start()
thread_2.start()
        
````

## Daemon Thread
Una thread **daemon** es una thread que acaba su ejecución si es que la thread principal termina. Por defecto, las threads no son **daemon**, por lo que tenemos que esperar que terminen para terminar el programa principal.

## Lock

Cuando queremos que una única thread esté en una sección de código a la vez debemos usar la clase `Lock`. Esto nos sirve cuando tenemos acceso a recursos compartidos. Debemos usar `lock.acquire()` para adquirir un lock y `lock.release()` para soltarlo.

````python
def sumador(lista):
    for i in range(10**6):
        lista[0] += int(1)
        # pregunta: por qué ponemos este int?

lista = [0]

thread_1 = Thread(target=sumador, args=(lista, ))
thread_2 = Thread(target=sumador, args=(lista, ))

thread_1.start()
thread_2.start() 

thread_1.join()
thread_2.join() 

print(lista)
# imprime cada vez algo distinto, watafak?¡
````

````python
from threading import Lock

def sumador(lista, lock):
    for i in range(10**6):
        lock.acquire()
        # en esta parte solo puede estar una línea a la vez
        lista[0] += int(1)
        lock.release()

lista = [0]
lock = Lock()

thread_1 = Thread(target=sumador, args=(lista, lock))
thread_2 = Thread(target=sumador, args=(lista, lock))

thread_1.start()
thread_2.start() 

thread_1.join()
thread_2.join() 

print(lista)
# ahora si :)
````

## Events
Con la clase **Event** podemos hace que una thread espere a que pase algún evento en otra thread, sin necesidad de que la otra thread termine.

````python
from threading import Event

class MiThread(Thread):

    def __init__(self, numero, evento):
        super().__init__()
        self.numero = numero
        self.evento = evento
        
    def run(self):
        print(f"Thread {self.numero}: Voy a contar!")

        if self.numero == 1:
            # la thread 1 espera a que la thread 2 cuente hasta 5
            self.evento.wait()

        for i in range(10):
            print(f"Thread {self.numero}: {i}")
            time.sleep(0.5 * self.numero)
            # cuando la thread 2 cuenta hasta 5, setea el evento
            if self.numero == 2 and i == 5:
                self.evento.set()
        print(f"Thread {self.numero}: Terminé :v")


evento = Event()
thread_1 = MiThread(1, evento)
thread_2 = MiThread(2, evento)

thread_1.start()
thread_2.start()
        
````

## Deadlocks
Los **deadlocks** se producen cuando ambas threads están esperando a que la otra haga algo, por lo que ninguna puede avanzar.

````python
from threading import Event

class MiThread(Thread):

    def __init__(self, numero, evento_1, evento_2):
        super().__init__()
        self.numero = numero
        self.evento_1 = evento_1
        self.evento_2 = evento_2
        
    def run(self):
        print(f"Thread {self.numero}: Voy a contar!")

        # la thread 1 espera al evento 2 para setear el evento 1
        if self.numero == 1:
            self.evento_2.wait()
            self.evento_1.set()

        # la thread 2 espera al evento 1 para setear el evento 2
        if self.numero == 2:
            self.evento_1.wait()
            self.evento_2.set()

        for i in range(10):
            print(f"Thread {self.numero}: {i}")
            time.sleep(0.5 * self.numero)
        print(f"Thread {self.numero}: Terminé :v")


evento_1 = Event()
evento_2 = Event()
thread_1 = MiThread(1, evento_1, evento_2)
thread_2 = MiThread(2, evento_1, evento_2)

thread_1.start()
thread_2.start()
        
````

# Interfaces Gráficas

Para programar interfaces gráficas, usamos la separación frontend/backend. El **frontend** contiene los elementos visuales: ventanas, botones, cuadros de texto, etc. El **backend** contiene la lógica detrás del funcionamiento de la interfaz: verificaciones, cálculos, etc.

Las interfaces gráficas buscan tener **alta cohesión** y **bajo acoplamiento**:

- **Cohesión**: cada una de las partes tiene una responsabilidad común y no hace cosas fuera de eso, por ejemplo, en el backend no revisamos si una contraseña está bien formateada.
- **Acoplamiento**: si cambiamos una de las partes, necesariamente tenemos que cambiar la otra, es decir, los módulos dependen mucho entre sí.

También podemos comparar una arquitectura lineal con la arquitectura basada en eventos que usamos en interfaces gráficas:

- **Arquitectura lineal**: las tareas se ejecutan secuencialmente, una tras otra. Esto puede ser ineficiente en sistemas interactivos, donde una operación bloqueante (como esperar un clic) detendría todo el programa.
- **Arquitectura basada en eventos**: las tareas se disparan en respuesta a eventos (como un clic o una tecla presionada). Esto permite ejecutar múltiples acciones de forma concurrente y no bloqueante, lo que es ideal para interfaces gráficas.

## Señales
Las señales nos permiten comunicación entre frontend y backend. El método `connect` conecta la señal a alguna función, y el método `emit` emite la señal. Las señales se pueden emitir con **argumentos**, estos deben coincidir con los argumentos que toma la función a la cual están conectadas.

  class Ventana_1(QWidget):

      contrasena = self.input.text()
      self.senal_contrasena.emit(contrasena)

      if aprobado:
        self.hide()
        self.senal_abrir_ventana.emit()

  class Ventana_2(QWidget):

      super().__init__()

  class Verificador(QObject):
    senal_respuesta = pyqtSignal(bool)

    def __init__(self):
      super().__init__()
    
    def verificar_contrasena(self, contrasena):
      if len(contrasena) < 8:
        self.senal_respuesta.emit(False)
      else:
        tiene_dig = False
        tiene_may = False
        tiene_min = False
        for caracter in contrasena:
          if caracter.isdigit():
            tiene_dig = True
          if caracter.isupper():
            tiene_may = True
          if caracter.islower():
            tiene_min = True
        if tiene_dig and tiene_may and tiene_min:
