# *Paths*

Un *path* (o ruta) nos indica un lugar donde se encuentra un archivo o un directorio (una carpeta) dentro de nuestro computador. El sistema de archivos y directorios es representado como un *árbol*.

![arbol](./assets/arbol_file_system.png)

Un ejemplo sería `Pedro/Fotos/foto1.jpg`, que indica que el archivo `foto1.jpg` se encuentra dentro del directorio `Fotos`, que a su vez se encuentra dentro del directorio `Pedro`.

Un *path* puede ser absoluto o relativo:

- Un *path* absoluto es la ruta a un archivo desde la **raíz** del sistema de archivos. Observando la figura, la raíz del sistema sería `/`, por lo que el *path* absoluto de el archivo `foto1.jpg` seria: `/Users/Pedro/Fotos/foto1.jpg`.
    - Ojo: En muchos sistemas operativos, la raíz se representa como `/`, pero en Windows lo puedes ver como `\`, o `C:/`.

- Un *path* relativo es la ruta a un archivo desde un cierto directorio. Este debe interpretarse a partir de algún directorio específico.

# La Terminal

La terminal (o consola) es una herramienta que nos permite interactuar con nuestro computador a través de **comandos**. A continuación se listan los comandos más comunes vistos en este curso:
- `pwd`: Imprime el *path* absoluto a la ubicación donde tenemos abierta la terminal. 
- `dir` o `ls`: Lista los archivos y carpetas en la ubicación donde tenemos abierta la terminal. 
- `cd`: Nos permite navegar entre directorios. Recibe un argumento: el directorio al que queremos acceder.
- `mkdir`: Nos permite crear un directorio en la ubicación donde tenemos abierta la terminal. Recibe un argumento: el nombre que queremos colocarle al directorio.
- `touch`: Nos permite crear un archivo en la ubicación donde tenemos abierta la terminal. Recibe un argumento: el nombre que queremos colocarle al archivo.
- `python`: Nos permite ejecutar el intérprete de Python desde la terminal. Tiene 2 formas de uso posibles:
    - Sin argumentos: Abre la terminal interactiva de Python, lo que nos permite ejecutar código línea por línea. 
    - Con 1 argumento: Podemos indicar el nombre, (o el *path*) de un archivo y Python ejecutará el intérprete sobre el archivo completo, línea por línea.

# Git

Git es un sistema distribuido de control de versiones ampliamente utilizado en el entorno de desarrollo de *software*. Nos permite trabajar con repositorios (carpeta con archivos) en nuestro computador local con repositorios en la nube (gracias a GitHub). 

Git **almacena todas las versiones que hemos hecho de un archivo, y que hayamos decidido almacenar**. Para almacenar una versión, utilizamos un *commit*, el cual es un registro de la evolución de un archivo o conjunto de archivos.

## Comandos básicos de Git

### `git clone`

Nos permite 'descargar' un repositorio en la nube (desde GitHub) y traerlo a nuestro computador, convirtiéndose en un repositorio local.

Uso en la terminal: `git clone <url-repositorio-en-la-nube>`. Debes ejecutar este comando en la ubicación en la que quieras almacenar el repositorio local.

### `git pull`

Nos permite 'descargar' o traer todos los *commits* que fueron creados en el repositorio en la nube desde que lo clonamos a nuestro computador, y que por lo tanto no se encuentran en nuestro repositorio local. En simples palabras, nos permite 'sincronizar' nuestro repositorio local con el repositorio de la nube actualizado.

Uso en la terminal: `git pull`.

### `git add`

Agrega archivos que hayamos modificado a una lista de cambios (area de *staging*).

Uso en la terminal: `git add <paths>`, donde `paths` son *paths* a archivos o carpetas en nuestro repositorio local que queremos agregar al area de *staging*, es decir, moverlos a nuestra lista de cambios. Si quieres agregar TODOS los cambios que hiciste en el repositorio local, puedes usar `git add .`.


### `git commit`

Crea un *commit* (o una versión de nuestros archivos) en nuestro repositorio local. Este registro:
- La hora a la que se hizo el *commit*
- Los cambios que se realizaron a archivos en el repositorio. Dichos cambios pueden ser:
    - Crear un archivo en el repositorio local
    - Modificar un archivo existente en el repositorio local
    - Eliminar un archivo existente en el repositorio local
- Metadata: Datos adicionales no relevantes para este curso

Uso en la terminal: `git commit -m "<mensaje>"`, donde `mensaje` es un mensaje descriptivo explicando lo que estás almacenando en esta versión de los archivos.

### `git push`

Publica y sube al repositorio remoto todas las versiones (o *commits*) que hayamos realizado en nuestro repositorio local y que aún no hayan sido subidas al repositorio remoto.

Uso en la terminal: `git push`.

### Para crear e instanciar una Namedtuple
El uso que se le da a una namedtuple es similar al de una clase sin métodos. Para definir la namedtuple, debes ingresar como argumentos el nombre que guardará python para el tipo de las instancias y el nombre de los atributos dentro de una lista:

```python
from collections import namedtuple

# DOS FORMAS DE CREAR LAS NAMEDTUPLES
Mascotas = namedtuple('mascotas_type', ['nombre', 'edad', 'animal'])
Duenos = namedtuple('duenos_type', 'nombre, nombre_mascota')

micky = Mascotas('Micky Andres', 7, 'Perro')
cristian = Duenos('Cristian Andres', 'Micky Andres')

print(micky)
print(cristian)

# NOTEN: el nombre que se usa para instanciar la namedtuple es el de la variable
# sin embargo python lo guarda con el str dado como primer argumento
```

### Inmutables
No puedes cambiar los atributos de una namedtuple una vez ya la instancias, pero puedes crear una copia con el atributo cambiado !!!

```python
micky.edad += 1
```

```python
micky_con_mas_edad = micky._replace(edad = micky.edad + 1)

# Como es una copia, la instancia Micky queda intacta, pero que pasaría 
# si guardara en micky la copia ???

print(micky)
print(micky_con_mas_edad)
```

### LIFO
Los stacks son estructuras de datos LIFO (Last In First Out) que como lo dice su nombre, significa que si miras o eliminas algun elemento del stack, será el último que haya sido agregado. Esto es comparable a una pila de objetos, no puedes llegar y quitar la base, pero si puedes mover las ultimas piezas agregadas !!!

### Modelación con listas en Python
Los stacks son estructuras mutables, secuenciales, basados en push (similar a append) y en pop (LIFO), lo que hace conveniente representarlos con listas en Python !!!

### Basados en Push y Pop
Tal como las listas de Python los stacks se basan en operaciones que agregan elementos al final del stack y quitan los elementos desde el final del stack, lo que puede ser modelado con el append y pop :D

```python
stack = []
stack.append(1)
stack.append(2)
stack.append(3)

print(stack.pop())
print(stack)
```

### Secuenciales
Los stacks tienen un orden determinado y fijo, ya que es importante para la correcta representación de "pila" de elementos, no quieres que de la nada la base de tu stack esté al medio ni que el último elemento agregado sea la base, por lo que es necesario que estos sean ordenados.

```python
stack = ['a', 'b', 'c', 'd']
stack.append('e')

for elemento in stack:
    print(elemento)
```

### Mutables
Parte de la representación de "pila" implica poder agregar o quitar elementos, lo que exige que la estructura sea mutable y lo hemos podido ver en los otros ejemplos donde modificamos los stacks definidos.

### Heterogeneidad
Pueden contener diferentes tipos de datos !!!

```python
stack = [1, 2.3, 'hola', (1, 2.3)] # No hay error !!!
```

### FIFO 
Las estructuras FIFO (First In First Out) como las queues son unas que como su nombre indica, el primer elemento agregado es también el primero en ser mirado o eliminado. Esta forma nos permite modelar situaciones como la fila de un supermercado, ya que te llegarían muchos reclamos por modelarlo con stacks y que el último cliente en llegar sea el primero atendido.

### Modelados con Listas y Deques en Python
Los queues también son modelados con listas, ya que podemos utilizar el metodo append y pop(0), pero como vimos en los contenidos esta última operación es MUY ineficiente, por lo que podemos utilizar los deques de la librería `collections`, estos son una doble cola, permitiéndote agregar y quitar elementos de las dos puntas del deque eficientemente.

### Basados en Enqueue y Dequeue
Para esta estructura de datos deseamos agregar elementos al final y retirar los del inicio, para ello podemos utilizar los metodos append y popleft() de los deques.

```python
from collections import deque
lista = []
deque = deque()

# ENQUEUE
lista.append(1)
lista.append(2)
lista.append(3)
lista.append(4)
print(f"La lista parte asi: {lista}")

deque.append(1)
deque.append(2)
deque.append(3)
deque.append(4)
print(f"el deque parte asi {deque}\n")

# DEQUEUE
elemento = lista.pop(0)
print(f"saque el elemento {elemento} y la lista luce asi: {lista}")

elemento2 = deque.popleft()
print(f"saque el elemento {elemento2} y el deque luce asi: {deque}")
```

### Secuencialidad
Los queues son estructuras secuenciales (ordenadas) por el mismo motivo que los stacks, su logica FIFO depende de esto, ya que no queremos que alguien que llegó último se cole en la fila del supermercado ni que cuando estemos primeros de la nada aparezcamos más atrás !!!

```python
from collections import deque
deque = deque([1, 2, 3, 4, 5, 6, 7])

for elemento in deque:
    print(elemento)
```

### Heterogeneidad
Pueden contener diferentes tipos de datos !!!

```python
from collections import deque
deque = deque([1, 2.3, 'hola', (1, 2)]) # No hay error !!!
```

Para poder agregar y eliminar elementos, las queues deben ser mutables tal como hemos visto en los ejemplos anteriores.

### Definición

Los diccionarios son estructuras de datos que permiten relacionar una llave con un valor, y estos se definen y funcionan asi:

```python
# creamos un dict que tiene de llave el string 'llave' y de valor el string 'valor'
diccionario = {'llave' : 'valor'}
```

```python
# Para acceder al valor de una llave, usamos la siguiente notación
print(diccionario['llave'])

# la notación previa genera errores cuando la llave llamada no existe !!!
# para probar descomenta la siguiente línea:
# print(diccionario['esta llave no existe'])
```

```python
# Para evitar los KeyErrors usamos el metodo get
# cuando la llave existe actúa igual que la primera notación
# y en el caso que no, retorna None !!!

valor_de_key_no_existente = diccionario.get('llave que no existe')
valor_de_key_existente = diccionario.get('llave')

print(valor_de_key_no_existente)
print(valor_de_key_existente)
```

### Mutable

Los diccionarios son mutables, por lo que puedes:
- Crear nuevos pares key/value
- Cambiarle el value a una key
- Borrar pares key value

A continuación tienes ejemplos de cada uno:

```python
# CREAR NUEVOS PARES KEY VALUE
# diccionario tiene 'llave' : 'valor'
# ahora crearé un nuevo par llave valor


# la llave de este par es 'otra llave' y su valor es la lista [67, 81]
diccionario['otra llave'] = [67, 81]


print(diccionario)
print(diccionario.get('otra llave'))

# NOTA QUE: esta notación nunca da error porque si la key no existe, la crea 
# y si existe, cambia su valor
```

```python
# Modificar el valor de una llave existente
# Cambiaré el valor de la key 'otra llave' a 'otro valor'

diccionario['otra llave'] = 'otro valor'

print(diccionario)
print(diccionario.get('otra llave'))
```

```python
# Para eliminar un par key value solo debemos usar la sentencia del
del diccionario['otra llave']

print(diccionario)
print(diccionario.get('otra llave'))

# CUIDADO CON LOS KEY ERROR, si la llave no existe, esto dara error
```

### Eficiencia
Los diccionarios tienen una busqueda de elementos eficiente, ya que toman tiempo constante, a diferencia de las listas, que necesitan ser recorridas hasta encontrar el valor buscado. Esto hace que la eficiencia de busqueda de los diccionarios no dependa de su largo.

Lo que hace que esto sea posible es el ***hash***, este es un numero que se le asigna a los objetos hasheables, es comparable a un id del elemento, pero no toda variable de python es ***hasheable***, para distinguir estos ten en cuenta:

1. Implementa el método __hash__. Este método retorna un entero, y sirve de entrada para la función de hash de la tabla de hash.
2. El valor que retorna __hash__ no cambia durante el ciclo de vida del objeto.
3. Implementa el método __eq__. Este método compara dos objetos y retorna True si éstos deben ser considerados iguales.
4. Si dos objetos son iguales según el método __eq__, entonces el valor que retorna __hash__ debe ser el mismo. No es necesario que esto se cumpla al revés.

Estas condiciones nos permiten notar que los objetos mutables NO serán hasheables.

```python
# Los dos strings tienen el mismo hash al ser iguales
str1 = 'HOLAAA'
str2 = 'HOLAAA'

print(hash(str1), hash(str2), str1 == str2)
```

```python
# las tuplas son inmutables mientras no contengan elementos mutables !!! 
# observa este caso
tupla1 = (1, 2, 3)
tupla2 = (1, [2, 3])

print(hash(tupla1))
print(hash(tupla2))
```

### Heterogeneidad (llaves y valores)

Que los diccionarios sean heterogéneos implica que tanto las llaves como los valores pueden ser de cualquier tipo de dato, pero las llaves pueden ser solo de tipos ***hasheables***

```python
diccionario = {
    'llave' : 123,
    123 : (1, 2, [3, 4]),
    (1, 2) : 'valor',
    3.14 : [1, 2, 3]}

# Aquí podemos ver ejemplos de un diccionario válido, donde todas sus llaves
# son hasheables, pero sus valores no necesariamente lo son !!!

print('diccionario', diccionario, '\n')

print(diccionario.get(123))
print(diccionario.get((1, 2)))
print(diccionario.get(3.14))
```

### Métodos útiles
Los diccionarios tienen los métodos:
- items(): retorna una tupla de tuplas de la forma ((llave, valor), (llave, valor), ...)
- keys(): retorna una tupla con todas las llaves del diccionario
- values(): retorna una tupla con todos los valores del diccionario

```python
diccionario = {
    'a' : 1,
    'b' : (1,2),
    'c' : 3.14,
    'd' : 'hola'
}

print(diccionario.items(), '\n')
print(diccionario.keys(), '\n')
print(diccionario.values(), '\n')
```

```python
# Se pueden recorrer, lo que los hace bastante útiles
for llave in diccionario.keys():
    print(llave)

print()

for valor in diccionario.values():
    print(valor)

print()

for llave, valor in diccionario.items():
    print(f'{llave} : {valor}')
```

### Diccionarios por comprensión
Puedes definir un diccionario sin necesidad de ingresar cada par llave valor si es que éstos tienen algun patrón, por ejemplo:

```python
n = 5
dobles = {i : 2 * i for i in range(n)}
print(dobles)
```

### Defaultdicts
- Definición
- Utilidad
- Cómo se inicializan los keys con los default values
Los defaultdicts son utiles para modelar cosas que tienen valores iniciales, por ejemplo:

```python
from collections import defaultdict

n = int(input())
contador_frutas = defaultdict(int) # inicializa en cero las nuevas llaves

for i in range(n):
    fruta = input()
    contador_frutas[fruta] += 1
```

```python
print(contador_frutas)
# podmos ver que incluso cuando no existia ninguna de estas llaves, no hubo
# un key error, ya que el defaultdict(int) las inicializa en 0
```

### Definición

Los sets son estructuras similares a un conjunto, es por esto que comparten características y operaciones. Su definición en python:

```python
set_no_vacio = {1 , 2 , 3}
set_vacio = set() 

print(set_no_vacio)
print(set_vacio)
print()

# no puedes usar los corchetes vacios porque estan reservados
# para diccionarios
print(type({}))
```

### Estructura sin elementos repetidos
También puedes definir sets desde una lista, tupla, stack, queue, etc. Si es que haces esto debes tener en cuenta que se pierden todos los elementos repetidos. 

Al igual que en un conjunto, el set solo indica si este elemento está o no, no cuántas veces pueda aparecer.

```python
lista = [1, 1, 3, 3, 1, 2, 1, 2, 3, 3, 2]
set_lista = set(lista)

print(set_lista)
```

### No secuenciales

Los sets no guardan de manera ordenada sus elementos, por lo que no permiten acceso por índice:

```python
print(set_lista[0])
```

###  Métodos

```python
reguetoneros = set()

# Para agregar un elemento al set
reguetoneros.add('Peachy Head')
reguetoneros.add('Rasputin Putin')
reguetoneros.add('Reo Bolsonaro')
reguetoneros.add('Callampin Bombin')
reguetoneros.add('Pololeador de Martes')

print(reguetoneros)
```

```python
# Para eliminar un elemento puedes usar remove o discard
reguetoneros.remove('Reo Bolsonaro')
print(reguetoneros)
```

```python
# remove puede darte un KeyError si el elemento no estaba en el set
# discard no hace nada en caso de que el elemento no esté y 
# lo borra si es que sí está
reguetoneros.discard('Rasputin Putin')
reguetoneros.discard('https:slashslash Bachelet')

# https:slashslash Bachelet NO estaba en el set y no da KeyError !!!
print(reguetoneros)
```

```python
# puedes verificar si un elemento está o no en un set con IN

if 'Pololeador de Martes' in reguetoneros:
    print('OMG, yo escucho a Pololeador de Martes')

if 'https:slashslash Bachelet' in reguetoneros:
    print('OMG, yo escucho a https:slashslash Bachelet')
```

#### Sub y Super Conjuntos en sets
Un set es subset de otro si el otro contiene todos sus elementos.
Un set es superset de otro si el otro es su subset.

```python
set_a = {1, 2, 3}
set_b = {0, 1, 2, 3}

print(f"A es subset de B:", set_a <= set_b)
print(f"A es superset de B:", set_a >= set_b)
print(f"A es igual a B:", set_a == set_b)
print()

print(f"B es subset de A:", set_b <= set_a)
print(f"B es superset de A:", set_b >= set_a)
print(f"B es igual a A:", set_b == set_a)
```

# Args y Kwargs

¿Cómo las funciones como `print` son capaces de recibir indefinidas cantidades de argumentos sin saber cuántos esperar ????? ARGS Y KWARGS !!!

Los args y kwargs son formatos diferentes de argumentos para una función que permiten utilizarlas con indefinidas cantidades de argumentos, ¿pero cuáles son sus diferencias??

Los args son los argumentos por posicion, son todos ellos que el programa asocia a una variable gracias a que son introducidos en orden. En cambio los kwargs son los argumentos que el programa asocia a una variable gracias a que los ingresaste junto al nombre de esta (keyword).

```python
def foo(arg1, arg2):
    print('arg1:', arg1)
    print('arg2:', arg2)
    print()

# argumentos por posición
foo(1, 2)

# argumentos por key
foo(arg2 = 3, arg1 = 6)
# nota que fueron indicados en desorden pero el programa identifica por
# la aclaración del arg2 = valor
```

Y si ahora quiero indeterminadas cantidades de argumentos posicionales? USEMOS ARGS!!!

```python
# Para usar args notamos que antes del nombre donde guardo los args agrego un *
# esto le indica al programa que recibiré una indefinida ctd de argumentos

def contar_e_imprimir(*argumentos_posicionales):
    print(argumentos_posicionales)

contar_e_imprimir(1,2,3,4,5,6,7) # LOS ARGS SON UNA TUPLA DE LOS VALORES INGRESADOS
```

```python
# Para usar kwargs notamos que antes del nombre donde guardo los args agrego **
# esto le indica al programa que recibiré una indefinida ctd de key word argumentos

def contar_e_imprimir(**argumentos_por_llave):
    print(argumentos_por_llave)

contar_e_imprimir(arg1 = 1, arg3 = 2, arg6 = 9.4) 
# LOS KWARGS SON UN DICCIONARIO DE LOS VALORES Y VARIABLES DADAS !!!
```

Pero hay reglas para su uso, cosas a notar:
- No puedes usar args despues de kwargs, ya que pierde el sentido el orden
- No puedes tener dos variables de args ni kwargs
- Puedes tener argumentos normales ademas de args y kwargs mientras respetes el orden

```python
def foo(arg1, arg2, arg3, *posicionales, **key_words): # CORRECTO ORDEN
    print(arg1, arg2, arg3, posicionales, key_words)

foo(1, 2, 3, 8, 9, 10, cosa = 3, cosa2 = 4)
```

## Herencia
Las clases en python nos permiten modelar diferentes objetos con sus atributos y métodos, pero la herencia de clases nos ayuda además a modelar situaciones de especialización, en donde haya una clase base y clases especializadas a partir de ésta. A éstas se les llama superclase y subclase respectivamente.

```python
class Animal: # SUPERCLASE
    def __init__(self, nombre):
        self.nombre = nombre

    def comer(self):
        print('ñam')

class Perro(Animal): # SUBCLASE DE ANIMAL
    def buscar_pelota(self):
        print(':p')

class Gato(Animal): # SUBCLASE DE ANIMAL
    def lavarse(self):
        print('lick lick')

gruñon = Gato('Gruñon')
micky = Perro('Micky Andres')

gruñon.comer() # ambos tienen el método heredado de animal
micky.comer()

print()

gruñon.lavarse() 
micky.buscar_pelota()
# cada uno tiene el metodo de su clase especializada
# pero perro micky no puede lavarse ni gato gruñon buscar pelota
# DESCOMENTA LAS LINEAS, DARAN ERROR !!!
# micky.lavarse()
# gruñon.buscar_pelota()
```

Si quisieras agregar atributos especiales para cada clase hija debes sobreescribir el metodo init, pero heredando los atributos que ya habían sido definidos en la clase padre.

```python
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def comer(self):
        print('ñam')

class Perro(Animal):
    def buscar_pelota(self):
        print(':p')

class Gato(Animal):
    def __init__(self, nombre): # SOBRESCRIBIR INIT !!! ------------------------
        super().__init__(nombre)
        self.vidas_restantes = 7

    def morir(self):
        self.vidas_restantes -= 1

    def lavarse(self):
        print('lick lick')

gatito = Gato('gruñon')
perrito = Perro('Micky Andres')

print(gatito.vidas_restantes)
# print(perrito.vidas_restantes) # DARA ERROR
```

## Polimorfismo
Cuando dos subclases comparten un método con el mismo nombre, python distingue a cuál llamar dependiendo de la instancia desde que la llames !!! Observa los métodos `hablar` de `Perro` y `Gato`:

```python
# EJEMPLO !!!
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def comer(self):
        print('ñam')

    def hablar(self):
        pass

class Perro(Animal):
    def hablar(self):
        print('guau guau')

class Gato(Animal):
    def hablar(self):
        print('miau miau')




micky = Perro('Micky Andres')
rory = Gato('Horrorosa')

micky.comer() # metodo comer en comun
rory.comer()
print()
micky.hablar() # metodos
rory.hablar()
```

## Multiherencia
Hay casos donde es necesario hacer árboles de herencia, como:
- Heredar de más de una clase padre
- Heredar de una clase que hereda de otra
- Heredar de dos clases que heredaban de una única

A continuación ejemplos de esto:

```python
# EJEMPLO !!!
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def comer(self):
        print('ñam')

    def hablar(self):
        pass

class Perro(Animal):
    def hablar(self):
        print('guau guau')

class GoldenRetriever(Perro): # Animal -> Perro -> GoldenRetriever
    def botar_pelo():
        print('brush')
```

```python
# EJEMPLO !!!
class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def comer(self):
        print('ñam')

    def hablar(self):
        pass

class Macota:
    def __init__(self):
        self.nombre_dueno = ''
    
    def adoptar(self, nuevo_dueno):
        self.nombre_dueno = nuevo_dueno

class Perro(Animal, Macota): # hereda de Animal y Mascota
    def hablar(self):
        print('guau guau')
```

La herencia simple sigue una unica línea de herencia como en el primer ejemplo, en cambio en el otro el arbol de herencia varía. 

Esto permite reutilizar código, tener más orden y representar jerarquías. Sin embargo pueden ocurrir conflictos en donde no se sabe que método utilizar en caso de haber dos con el mismo nombre, pero python soluciona esto con el orden de herencia, por ejemplo en el codigo anterior se heredaría primero animal y luego mascota.

## Clases Abstractas
Este tipo de clases no están hechas para ser instanciadas, sino para modelar una base para otras clases. Para ser definida como abstracta una clase debe heredar de `ABC`, que hace que no pueda ser instanciada directamente, además debe tener al menos un método abstracto (pero no es necesario que todos sus métodos lo sean) y sus clases hijas deben implementar todos los métodos abstractos que tenga.

Este modelo establece un mínimo de lo que debe tener una clase hija manteniendo los beneficios de la herencia común.

```python
from abc import ABC, abstractmethod

class AparatoMovil(ABC):
    def __init__(self):
        self.app = 'Home'

    @abstractmethod
    def abrir_app(self, app):
        pass

class Celular(AparatoMovil):
    def __init__(self, numero_telefono):
        self.numero = numero_telefono
        self.en_llamada_con = None
    
    def llamar(self, numero):
        self.en_llamada_con = numero

    def cortar(self):
        self.en_llamada_con = None

    def abrir_app(self, app):
        if not self.en_llamada_con:
            self.app = app
        else:
            print('No puedes hacer esto')

class Tablet(AparatoMovil):
    def abrir_app(self, app):
        self.app = app
        self.apuntes = {}

    def tomar_apuntes(self, ramo, apuntes):
        if self.app == 'OneNote':
            if self.apuntes.get(ramo):
                self.apuntes[ramo].append = apuntes
            else:
                self.apuntes[ramo] = [apuntes]

# INTENTA COMENTAR EL METODO ABRIR APP DE ALGUNA DE LAS DOS CLASES HIJAS
# E INTENTA INSTANCIAR LUEGO DE ESO
celu = Celular('+56 9 1234 5678')
tablet = Tablet()
```

## Decoradores
Los decoradores son funciones que reciben otra función y complementan su algoritmo con elementos extras. Esto es posible dado que las funciones en python son objetos de primera clase, lo que permite que una función reciba otra como argumento y la decore.

```python
def un_decorador(func):
    def cosa_extra():
        print('esto es una funcion que solo existe dentro del uso de la funcion',
        'un_decorador')
        func()
        print('estoy imprimiento desde cosa_extra')
        print('decoro esto como quiero')
        print('incluso puedo llamar de nuevo la funcion')
        func()
    
    # ahora dentro de la funcion un_decorador puedo pasar la funcion completa interna
    # como valor y se correra esto en vez de la funcion decorada, ya que esto la llama
    return cosa_extra

@un_decorador
def imprimir_HOLA():
    print('HOLA')

imprimir_HOLA()
```

## Properties: @property
Es un decorador que permite definir métodos en una clase que se comporten como atributos pero verifican condiciones y/o modifican de alguna forma el valor del verdadero atributo que puede ser privado.

```python
# Como la clase tiene el atributo de probabilidad ruleta, queremos que esté entre
# 0 y 1, ademas de ser un float, por ello creamos la property de este atributo
# guardamos el verdadero atributo como privado y definimos el setter y getter.
# En el setter cambiamos el valor solo si cumple con estar entre 0 y 1 y ser float

class Casino:
    def __init__(self):
        self._probabilidad_ruleta = 0

    @property
    def probabilidad_ruleta(self):
        return self._probabilidad_ruleta

    @probabilidad_ruleta.setter
    def probabilidad_ruleta(self, value):
        if (0 <= value <= 1) and isinstance(value, float):
            self._probabilidad_ruleta = value
```
