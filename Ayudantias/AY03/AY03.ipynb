## Pseudocódigo
El **pseudocódigo** es una forma de describir algoritmos utilizando un lenguaje cercano al humano, pero con una estructura similar a la de un lenguaje de programación. No sigue reglas sintácticas estrictas, sino que sirve para plasmar la lógica de un programa de manera clara antes de implementarlo en un lenguaje específico. Algunas ventajas del psudocódigo son:

* Ayuda a planificar la lógica sin preocuparse por la sintaxis exacta
* Facilita la comunicación de ideas entre programadores o equipos
* Permite detectar errores de diseño antes de codificar


### Ejemplo: división por cero
Vamos a crear un programa que intente dividir dos números ingresados por el usuario. Si hay un error de división por cero, lo manejaremos. En pseudocodigo tenemos:

```
Intentar:
    Pedir al usuario que ingrese el numerador (num1)
    Pedir al usuario que ingrese el denominador (num2)
    resultado = num1 / num2
    
Si hay un error de division por cero:
    Mostrar mensaje: "Error: división por cero."
    
Si no hay errores:
    Mostrar mensaje: "El resultado es: " + resultado
    
Finalmente:
    Mostrar mensaje: "Proceso terminado."
```

````python
try: # intenta ejecutar codigo que podría generar Errores
    
    num1 = float(input("Ingresa el numerador: "))
    num2 = float(input("Ingresa el denominador: "))
    resultado = num1 / num2

except ZeroDivisionError:  # Captura y maneja errores especificos
    
    print("Error: Division por cero.")

else:  # (opcional) se ejecuta solo si no hubo errores
    
    print(f"El resultado es: {resultado}")

finally: # (opcional) Siempre se ejecuta, haya errores o no
    
    print("Proceso terminado.")
````

# Codigo Base

Primero tenemos nuestra clase `estrella` que representará los cuerpos celestes que debemos explorar:

````python
class Estrella:
    def __init__(self, nombre, alias, magnitud, distancia, temperatura, radio,
                 luminosidad):
        self.nombre = nombre
        self.alias = alias
        self.magnitud = magnitud
        self.temperatura = temperatura
        self.distancia = distancia
        self.radio = radio
        self.luminosidad = luminosidad
````

Tambien, utilizaremos una función para leer las estrellas de nuestro `.csv` e instanciar los objetos, despues retornamos las instancias creadas dentro de un diccionario, donde la llave será el nombre de la estrella y el valor el objeto.

````python
def cargar_estrellas(nombre_archivo):
    with open(nombre_archivo, "rt", encoding="utf-8") as archivo:
        next(archivo) #Es un iterador :D
        texto = archivo.readlines()

    diccionario_estrellas = {}
    for linea in texto:
        nombre, alias, magnitud, distancia, temperatura, radio, luminosidad = linea.split(',')
        estrella = Estrella(
            nombre, alias, magnitud, float(distancia), float(temperatura),
            float(radio), float(luminosidad)
        )
        diccionario_estrellas[estrella.nombre] = estrella

    return diccionario_estrellas
````

Finalmente, utilizaremos otra función para cargar el nombre de las estrellas cercanas que debemos recorrer y retornamos una lista con los nombres.

````python
def cargar_nombres_estrellas_cercanas(nombre_archivo):
    with open(nombre_archivo, 'rt') as archivo:
        nombres = archivo.readlines()

    # Lo siguiente es una lista por comprensión
    return [nombre.strip() for nombre in nombres]
````

## Parte 1

Tenemos varias posibles fuentes de errores y debemos crear excepciones para cada tipo de error.


### Problema 1
Tenemos que el Alias de las estrellas tiene dos letras mayúsculas y dos números, siempre juntos (sólo puede haber 12AB o AB12), pero para que el alias sea correcto, las letras deben estar antes de los números y no puede contener la letra F. En caso de que alguno no se cumpla levantamos un error que contenga el siguiente mensaje:

`"Error: El alias de la estrella es incorrecto."`

````python
def verificar_alias_estrella(estrella):
    if estrella.alias[0:2].isnumeric() or "F" in estrella.alias:
        """
        Opciones:
        A) raise ValueError("Error: El alias de la estrella es incorrecto.")
        B) raise TypeError("Error: El alias de la estrella es incorrecto.")
        C) raise ZeroDivisionError("Error: El alias de la estrella es incorrecto.")
        D) raise NameError("Error: El alias de la estrella es incorrecto.")
        E) raise KeyError("Error: El alias de la estrella es incorrecto.")    
        """
        raise ValueError("Error: El alias de la estrella es incorrecto.")



````

Con este error identificado, podemos arreglar los alias incorrectos de las estrellas:
1) Cambiando las letras F por T.
2) Reordenando los números.
3) Imprimimos el siguiente mensaje

`f"El alias de {estrella.nombre} fue correctamente corregido.\n"`

````python
def corregir_alias_estrella(estrella):
    # Sabemos que en este bloque puede llegar a existir un error
    # Por lo tanto lo dejamos dentro de un try
    try:
        verificar_alias_estrella(estrella)

    # Nosotros tenemos pensado que puede haber un ValueError, así
    # que cuando suceda podemos corregir el error
    except ValueError as error:
        print(error)
        # 1)
        if "F" in estrella.alias:
            estrella.alias = estrella.alias.replace("F", "T")
        # 2) 
        if estrella.alias[0:2].isnumeric():
            alias = estrella.alias[2:4]
            alias += estrella.alias[0:2]
        # 3)
        print(f"El alias de {estrella.nombre} fue correctamente corregido.\n")
````

### Problema 2

Debido a la colisión, las distancias de algunas estrellas se volvieron *negativas*, por lo que debemos identificar este error e informarlo

````python
# esta funcion permite levanta la Excepcion
def verificar_distancia_estrella(estrella):
    if estrella.distancia < 0:
        raise ValueError("Error: Distancia negativa.")
````

Ademas crea una función capaz de:
* Volver positiva la distancia en caso de ser negativa
* Si no levantó ningún error se debe sumar 10 a la distancia con la estrella
* En cualquier caso se debe **Siempre** imprimir la distancia final

````python
# Esta función captura la excepcion y corrige el formato
def corregir_distancia_estrella(estrella):
    try:
        verificar_distancia_estrella(estrella)
    except ValueError as error:
        print(error)
        estrella.distancia = -1 * estrella.distancia
        print(f"La distancia de la estrella {estrella.nombre} fue corregida.")
    else:
        estrella.distancia += 10
    finally:
        print(f"La estrella {estrella.nombre} se encuentra a una distancia real de {estrella.distancia}\n")
````

### Problema 3

3. Otro problema que proviene de las colisiones, es que las magnitudes de las estrellas cambiaron de tipo y hasta algunas ocupan ";" para dividir la parte entera de la fracción del número!

3.14 -----> 3;14 😲

De nuevo necesitamos crear una función que identifique el error y otra que lo arregle

````python
# esta funcion permite levanta la Excepcion
def verificar_magnitud_estrella(estrella):
    if not isinstance(estrella.magnitud, float):
        """
        Opciones:
        A) raise ZeroDivisionError("Error: Magnitud no es del tipo correcto.")
        B) raise AttributeError("Error: Magnitud no es del tipo correcto.")
        C) raise KeyError("Error: Magnitud no es del tipo correcto.")
        D) raise TypeError("Error: Magnitud no es del tipo correcto.")
        """
        raise TypeError("Error: Magnitud no es del tipo correcto.")
        pass


# Esta función captura la excepcion y corrige el formato
def corregir_magnitud_estrella(estrella):
    try:
        verificar_magnitud_estrella(estrella)
    except TypeError as error:
        print(error)
        if ";" in estrella.magnitud:
            estrella.magnitud = float(estrella.magnitud.replace(";", "."))
        estrella.magnitud = float(estrella.magnitud)
        print(f"La magnitud de la estrella {estrella.nombre} fue corregida.\n")
````

5. Finalmente utilizaremos un diccionario para guardar la información de las estrellas, usando la llave con el nombre de la estrella. Por lo que en caso de no existir la llave, debemos levantar una excepción para indicarlo.

````python
def dar_alerta_estrella_cercana(nombre_estrella, diccionario_estrellas):
    try:
        estrella = diccionario_estrellas[nombre_estrella]
        print(
            f"Estrella {nombre_estrella} está en nuestra base de datos." +
            f" Su alias es {estrella.alias}."
        )
        """"
        Opciones:
        A) except ValueError:
        B) except TypeError:
        C) except IdentationError:
        D) except NameError:
        E) except KeyError:
        """
    except KeyError:
        print(
            f"Estrella {nombre_estrella} NO está en nuestra base de datos." +
            " Alerta!, puede ser una trampa de algún extraterrestre 👽!"
        )
````

# Parte 2: Ruta peligrosa

Ahora debemos deterrmianr una ruta para nuestro recorrido, esto quiere decir que debemos comprobar si los valores de luminocidad, magnitud y temperatura suponen un peligro para nuestra nave, para esto debemos levantar excepciones en caso de ser necesario.

````python
def verificar_condiciones_estrella(estrella):
    if estrella.luminosidad > 15500:
        raise ValueError(f"{estrella.nombre}: Luz excesiva")
    elif estrella.magnitud > 4:
        raise ValueError(f"{estrella.nombre}: Estrella excesivamente grande")
    elif estrella.temperatura > 7200:
        raise ValueError(f"{estrella.nombre}: Estrella muy caliente")
````

3. Por ultimo, tendremos una función que recibe una lista de Estrellas, esta descartara aquellas peligrosas y retornara una lista con las estrellas seguras. Es esta función la que implementa los bloques `try/except` desde donde levantaremos la expceción personalizada.

````python
def generar_ruta_estrellas(estrellas):
    ruta_de_estrellas = []
    for estrella in estrellas:
        try:
            verificar_condiciones_estrella(estrella)
            ruta_de_estrellas.append(estrella.nombre)
            print(f'¡La Estrella {estrella.nombre} se ha agregado a tu ruta!'
             + u'\x02' + '\n')

        except ValueError as error:
            print(f'Alerta {error}')
            print("Esta estrella tuvo que ser eliminada de la ruta")

    return ruta_de_estrellas
````

# Analizar la ruta

Con todo el código finalizado, abrimos los archivos, arreglamos los errores que se crearon y calculamos la ruta de estrellas segura.

````python
def realizar_analisis(archivo_estrellas="Data/estrellas.csv", archivo_estrellas_cercanas= "Data/estrellas_cercanas.txt"):
    # Cargamos los archivos
    diccionario_estrellas = cargar_estrellas(archivo_estrellas)
    nombres_estrellas = cargar_nombres_estrellas_cercanas(archivo_estrellas_cercanas)

    print("Revisando posibles errores en las estrellas...\n")

    # Realizamos la comprobación de que todas las estrellas tengan tipos válidos
    for estrella in diccionario_estrellas.values():
        corregir_alias_estrella(estrella)
        corregir_distancia_estrella(estrella)
        corregir_magnitud_estrella(estrella)
    
    # Revisamos que estrellas son cercanas
    print("Revisando estrellas inexistentes...\n")
    for nombre_estrella in nombres_estrellas:
        dar_alerta_estrella_cercana(nombre_estrella, diccionario_estrellas)

    # Obtenemos la mejor ruta
    print("Generando ruta de estrellas...\n")
    ruta_de_estrellas = generar_ruta_estrellas(diccionario_estrellas.values())

    # Hacemos unos prints muy bonitos :) 
    print(58*u'\xb7')
    print('¡Enhorabuena! ¡Te has salvado de las estrellas peligrosas!\n')
    print(u'\u2554' + 5*u'\u2550' + ' RUTA DE ESTRELLAS ' + 5*u'\u2550' + u'\u2557')
    for estrella in ruta_de_estrellas:
        dif = 29 - len(estrella)
        if dif % 2 == 0:
            espacio1 = int(dif/2)
            espacio2 = int(dif/2)
        else:
            espacio1 = dif//2
            espacio2 = (dif//2) + 1
        print(u'\u2551' + espacio1*' ' + estrella + espacio2*' ' + u'\u2551')
    print(u'\u255a' + 29*u'\u2550' + u'\u255d')
````

# Archivos

### Personas

* Nombre
* Contraseña

# Definición Persona

````python
from collections import namedtuple

Persona = namedtuple('Persona', ['nombre', 'contrasena'])
````

# Personalización de errores

Como se mencionó, necesitamos crear nuestros propios errores, por lo que creamos clases que heredan de `Exception`.

````python
class ErrorNoAutorizado(Exception):
    def __init__(self, nombre):
        super().__init__(f"No pasarás, {nombre}, sucio traidor!")

class ErrorContrasenaIncorrecta(Exception):
    def __init__(self, nombre):
        super().__init__(f"Creo que escribiste mal tu contraseña, {nombre}")
````

# Funciones

Para que funcionen correctamente nuestros errores, vamos a implementar funciones que manejen el levantamiento de las excepciones.

````python
def verificar_usuario(nombre: str, bdd: list):
    for persona in bdd:
        if persona.nombre == nombre:
            return persona
    raise ErrorNoAutorizado(nombre)

def verificar_contrasena(persona: Persona, contrasena: str):
    if contrasena != persona.contrasena:
        raise ErrorContrasenaIncorrecta(persona.nombre)
````

# DCConexionSegura

Ahora necesitamos nuestro ente que controle la verificación de usuarios.

````python
class DCConexionSegura:
    def __init__(self, personas: list):
        self.personas = personas
    
    def autenticar(self, nombre: str, contrasena: str):
        try:
            persona = verificar_usuario(nombre, self.personas) 
            verificar_contrasena(persona, contrasena)
            print(f"Bienvenid@, {nombre}!")
        
        except ErrorNoAutorizado as error:
            print(error)
        
        except ErrorContrasenaIncorrecta as error:
            print(error)
````

Finalmente debemos implementar una función que cargue los datos desde el archivo `personas.csv`.

````python
def cargar_personas(ruta) -> list:
    personas = []
    with open(ruta, "rt") as archivo:
        archivo.readline()
        linea = archivo.readline()
        while linea:
            nombre, contrasena = linea.strip().split(",")
            personas.append(Persona(nombre,contrasena))
            linea = archivo.readline()
    return personas
````

Ahora sí podemos observar el comportamiento de nuestro código.
